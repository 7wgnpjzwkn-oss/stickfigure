```
import bpy
import math
import mathutils
import os

# ==============================================================================
# 1. ENVIRONMENT SETUP, SCENE CLEANING & RENDER OPTIMIZATIONS
# ==============================================================================
if bpy.context.object and bpy.context.object.mode != 'OBJECT':
    bpy.ops.object.mode_set(mode='OBJECT')

bpy.ops.object.select_all(action='SELECT')
bpy.ops.object.delete()

scene = bpy.context.scene
scene.frame_start = 1
scene.frame_end = 160
scene.frame_current = 1
scene.render.engine = 'CYCLES'

# --- CYCLES SPEED OPTIMIZATIONS ---
# Enable GPU rendering if available
preferences = bpy.context.preferences
addons = preferences.addons
if "cycles" in addons:
    cpref = preferences.addons['cycles'].preferences
    cpref.refresh_devices()
    # Attempt to use CUDA or OPTIX if available, fallback to CPU gracefully
    for device_type in ('OPTIX', 'CUDA', 'OPENCL'):
        try:
            cpref.compute_device_type = device_type
            scene.cycles.device = 'GPU'
            break
        except TypeError:
            pass

# Set aggressive but clean sampling boundaries
scene.cycles.use_adaptive_sampling = True
scene.cycles.adaptive_threshold = 0.05  # Standard threshold for preview/clean renders
scene.cycles.samples = 64                # Cap max samples (factory default is often 4096)
scene.cycles.preview_samples = 16

# Limit max light paths (Massive speedup for outdoor scenes)
scene.cycles.max_bounces = 6
scene.cycles.diffuse_bounces = 4
scene.cycles.glossy_bounces = 4
scene.cycles.transmission_bounces = 2
scene.cycles.transparent_max_bounces = 4
scene.cycles.volume_bounces = 0

# Persistent Data keeps geometry cached in memory between frames
scene.render.use_persistent_data = True

# ==============================================================================
# 2. CAMERA INITIALIZATION
# ==============================================================================
bpy.ops.object.camera_add(location=(0, -12, 2.5), rotation=(math.radians(84), 0, 0))
camera = bpy.context.object
scene.camera = camera

# ==============================================================================
# 3. LIGHT RIG CALIBRATION
# ==============================================================================
bpy.ops.object.select_all(action='DESELECT')
for obj in bpy.data.objects:
    if obj.type == 'LIGHT':
        obj.select_set(True)
bpy.ops.object.delete()

sun_data = bpy.data.lights.new(name="Key_Sun", type='SUN')
sun_data.energy = 2.2
sun_data.angle = 0.15
sun_data.color = (1.0, 0.96, 0.91)
sun_obj = bpy.data.objects.new(name="Key_Sun", object_data=sun_data)
bpy.context.collection.objects.link(sun_obj)
sun_obj.location = (12, -12, 18)
sun_obj.rotation_euler = (0.78, 0.52, 0.65)

fill_data = bpy.data.lights.new(name="Sky_Fill", type='SUN')
fill_data.energy = 0.45
fill_data.color = (0.82, 0.88, 1.0)
fill_obj = bpy.data.objects.new(name="Sky_Fill", object_data=fill_data)
bpy.context.collection.objects.link(fill_obj)
fill_obj.rotation_euler = (-0.6, -0.4, 0.0)

if bpy.context.scene.world:
    world_nodes = bpy.context.scene.world.node_tree.nodes
    bg_node = next((n for n in world_nodes if n.type == 'BACKGROUND'), None)
    if bg_node:
        bg_node.inputs['Strength'].default_value = 0.75
        bg_node.inputs['Color'].default_value = (0.45, 0.55, 0.7, 1.0)

bpy.context.scene.view_settings.view_transform = 'AgX'
bpy.context.scene.view_settings.look = 'AgX - Punchy'
bpy.context.scene.view_settings.exposure = -3.5

# ==============================================================================
# 4. PROCEDURAL SKY MATERIAL SYSTEM
# ==============================================================================
world = scene.world
world.use_nodes = True
wnodes = world.node_tree.nodes
wlinks = world.node_tree.links
wnodes.clear()

node_w_output = wnodes.new(type='ShaderNodeOutputWorld')
node_w_background = wnodes.new(type='ShaderNodeBackground')
node_w_sky = wnodes.new(type='ShaderNodeTexSky')

node_w_sky.sky_type = 'MULTIPLE_SCATTERING'
node_w_sky.sun_elevation = math.radians(16)
node_w_sky.sun_rotation = math.radians(130)
node_w_sky.altitude = 50.0
node_w_sky.air_density = 1.0
node_w_sky.aerosol_density = 0.4
node_w_sky.ozone_density = 2.0

wlinks.new(node_w_sky.outputs['Color'], node_w_background.inputs['Color'])
wlinks.new(node_w_background.outputs['Background'], node_w_output.inputs['Surface'])

# ==============================================================================
# 5. SHADER MATERIAL INITIALIZATION (GRASS & BOULDERS)
# ==============================================================================
grass_mat = bpy.data.materials.new(name="Grass_Material")
grass_mat.use_nodes = True
g_nodes = grass_mat.node_tree.nodes
g_links = grass_mat.node_tree.links
g_nodes.clear()

node_g_output = g_nodes.new(type='ShaderNodeOutputMaterial')
node_g_principled = g_nodes.new(type='ShaderNodeBsdfPrincipled')
node_g_principled.inputs['Base Color'].default_value = (0.12, 0.38, 0.08, 1.0)
node_g_principled.inputs['Roughness'].default_value = 0.6

if 'Subsurface Weight' in node_g_principled.inputs:
    node_g_principled.inputs['Subsurface Weight'].default_value = 0.25
elif 'Subsurface' in node_g_principled.inputs:
    node_g_principled.inputs['Subsurface'].default_value = 0.25

g_links.new(node_g_principled.outputs['BSDF'], node_g_output.inputs['Surface'])

boulder_mat = bpy.data.materials.new(name="Boulder_Material")
boulder_mat.use_nodes = True
b_nodes = boulder_mat.node_tree.nodes
b_links = boulder_mat.node_tree.links
b_nodes.clear()

node_b_output = b_nodes.new(type='ShaderNodeOutputMaterial')
node_b_principled = b_nodes.new(type='ShaderNodeBsdfPrincipled')
node_b_principled.inputs['Roughness'].default_value = 0.95

node_b_coord = b_nodes.new(type='ShaderNodeTexCoord')
node_b_mapping = b_nodes.new(type='ShaderNodeMapping')
node_b_mapping.inputs['Scale'].default_value = (2.0, 2.0, 2.0)
b_links.new(node_b_coord.outputs['Generated'], node_b_mapping.inputs['Vector'])

node_b_color_noise = b_nodes.new(type='ShaderNodeTexNoise')
node_b_color_noise.inputs['Scale'].default_value = 3.0
node_b_color_noise.inputs['Detail'].default_value = 4.0  # Reduced from 6.0
b_links.new(node_b_mapping.outputs['Vector'], node_b_color_noise.inputs['Vector'])

node_b_ramp = b_nodes.new(type='ShaderNodeValToRGB')
node_b_ramp.color_ramp.elements[0].position = 0.3
node_b_ramp.color_ramp.elements[0].color = (0.28, 0.24, 0.21, 1.0)
node_b_ramp.color_ramp.elements[1].position = 0.7
node_b_ramp.color_ramp.elements[1].color = (0.42, 0.44, 0.45, 1.0)
b_links.new(node_b_color_noise.outputs['Color'], node_b_ramp.inputs['Fac'])
b_links.new(node_b_ramp.outputs['Color'], node_b_principled.inputs['Base Color'])

node_b_bump_noise = b_nodes.new(type='ShaderNodeTexNoise')
node_b_bump_noise.inputs['Scale'].default_value = 24.0
node_b_bump_noise.inputs['Detail'].default_value = 4.0  # Reduced from 8.0
b_links.new(node_b_mapping.outputs['Vector'], node_b_bump_noise.inputs['Vector'])

node_b_bump = b_nodes.new(type='ShaderNodeBump')
node_b_bump.inputs['Strength'].default_value = 0.45
node_b_bump.inputs['Distance'].default_value = 0.05
b_links.new(node_b_bump_noise.outputs['Fac'], node_b_bump.inputs['Height'])
b_links.new(node_b_bump.outputs['Normal'], node_b_principled.inputs['Normal'])
b_links.new(node_b_principled.outputs['BSDF'], node_b_output.inputs['Surface'])

# ==============================================================================
# 6. PROCEDURAL GRAVEL ROAD SHADER
# ==============================================================================
road_mat = bpy.data.materials.new(name="Gravel_Road_Material")
road_mat.use_nodes = True
nodes = road_mat.node_tree.nodes
links = road_mat.node_tree.links
nodes.clear()

output_node = nodes.new(type='ShaderNodeOutputMaterial')
output_node.location = (800, 300)
principled_node = nodes.new(type='ShaderNodeBsdfPrincipled')
principled_node.location = (500, 300)
links.new(principled_node.outputs['BSDF'], output_node.inputs['Surface'])

tex_coord = nodes.new(type='ShaderNodeTexCoord')
tex_coord.location = (-1200, 300)

mapping = nodes.new(type='ShaderNodeMapping')
mapping.location = (-1000, 300)
mapping.inputs['Scale'].default_value = (600.0, 5.0, 1.0)
links.new(tex_coord.outputs['Generated'], mapping.inputs['Vector'])

soil_noise = nodes.new(type='ShaderNodeTexNoise')
soil_noise.location = (-700, 500)
soil_noise.inputs['Scale'].default_value = 4.0
soil_noise.inputs['Detail'].default_value = 4.0  # Reduced from 8.0
links.new(mapping.outputs['Vector'], soil_noise.inputs['Vector'])

soil_ramp = nodes.new(type='ShaderNodeValToRGB')
soil_ramp.location = (-450, 500)
soil_ramp.color_ramp.elements[0].color = (0.16, 0.11, 0.08, 1.0)
soil_ramp.color_ramp.elements[1].color = (0.35, 0.28, 0.22, 1.0)
links.new(soil_noise.outputs['Color'], soil_ramp.inputs['Fac'])

gravel_voronoi = nodes.new(type='ShaderNodeTexVoronoi')
gravel_voronoi.location = (-700, 150)
gravel_voronoi.distance = 'EUCLIDEAN'
gravel_voronoi.feature = 'F1'
gravel_voronoi.inputs['Scale'].default_value = 85.0
gravel_voronoi.inputs['Randomness'].default_value = 0.9
links.new(mapping.outputs['Vector'], gravel_voronoi.inputs['Vector'])

stone_ramp = nodes.new(type='ShaderNodeValToRGB')
stone_ramp.location = (-450, 150)
stone_ramp.color_ramp.elements[0].color = (0.32, 0.32, 0.33, 1.0)
stone_ramp.color_ramp.elements[1].color = (0.58, 0.56, 0.54, 1.0)
links.new(gravel_voronoi.outputs['Color'], stone_ramp.inputs['Fac'])

mix_color = nodes.new(type='ShaderNodeMix')
mix_color.data_type = 'RGBA'
mix_color.location = (-150, 350)
mix_color.inputs['Factor'].default_value = 0.65
links.new(soil_ramp.outputs['Color'], mix_color.inputs[6])
links.new(stone_ramp.outputs['Color'], mix_color.inputs[7])
links.new(mix_color.outputs['Result'], principled_node.inputs['Base Color'])

bump_node = nodes.new(type='ShaderNodeBump')
bump_node.location = (150, 50)
bump_node.inputs['Strength'].default_value = 0.90
bump_node.inputs['Distance'].default_value = 0.03
links.new(gravel_voronoi.outputs['Distance'], bump_node.inputs['Height'])
links.new(bump_node.outputs['Normal'], principled_node.inputs['Normal'])

rough_ramp = nodes.new(type='ShaderNodeValToRGB')
rough_ramp.location = (150, 300)
rough_ramp.color_ramp.elements[0].position = 0.2
rough_ramp.color_ramp.elements[0].color = (0.55, 0.55, 0.55, 1.0)
rough_ramp.color_ramp.elements[1].position = 0.8
rough_ramp.color_ramp.elements[1].color = (0.92, 0.92, 0.92, 1.0)
links.new(gravel_voronoi.outputs['Distance'], rough_ramp.inputs['Fac'])
links.new(rough_ramp.outputs['Color'], principled_node.inputs['Roughness'])

road_displace_tex = bpy.data.textures.new(name="Road_Macro_Deform", type='CLOUDS')
road_displace_tex.noise_scale = 0.12

# ==============================================================================
# 7. OPTIMIZED LOW-POLY GRASS CARD ASSET
# ==============================================================================
verts = [
    (-0.04, 0.0, 0.0),
    (0.04, 0.0, 0.0),
    (-0.02, 0.0, 0.25),
    (0.02, 0.0, 0.25),
    (0.0, 0.03, 0.5)
]
faces = [
    (0, 1, 3, 2),
    (2, 3, 4)
]

mesh = bpy.data.meshes.new("Grass_Blade_Mesh")
mesh.from_pydata(verts, [], faces)
mesh.update()

blade = bpy.data.objects.new("Grass_Blade_Asset", mesh)
bpy.context.collection.objects.link(blade)
blade.data.materials.append(grass_mat)

blade.hide_render = True
blade.hide_viewport = True

# Terrain Elements
bpy.ops.mesh.primitive_plane_add(size=1000, location=(0, 0, -0.5))
ground = bpy.context.object
ground.name = "Grass_Terrain_Backdrop"
ground.data.materials.append(grass_mat)

# Optimizing grid size to save memory and parsing overhead
bpy.ops.mesh.primitive_grid_add(x_subdivisions=100, y_subdivisions=10, size=1.0, location=(0, 0, 0.1))
road = bpy.context.object
road.name = "Gravel_Road"
road.scale = (1000.0, 4.0, 1.0)
bpy.ops.object.transform_apply(scale=True, rotation=False, location=False)

road.data.materials.clear()
road.data.materials.append(road_mat)

road_displace = road.modifiers.new(name="Road_3D_Displacement", type='DISPLACE')
road_displace.texture = road_displace_tex
road_displace.strength = 0.08
road_displace.direction = 'Z'

# Foreground / Background Meadow Bounds
bpy.ops.mesh.primitive_plane_add(size=1, location=(0, -29.0, 0))
meadow_near = bpy.context.object
meadow_near.name = "Grass_Meadow_Near"
meadow_near.scale = (45, 54, 1)
bpy.ops.object.transform_apply(scale=True, rotation=False, location=False)
meadow_near.data.materials.append(grass_mat)

bpy.ops.mesh.primitive_plane_add(size=1, location=(0, 29.0, 0))
meadow_far = bpy.context.object
meadow_far.name = "Grass_Meadow_Far"
meadow_far.scale = (45, 54, 1)
bpy.ops.object.transform_apply(scale=True, rotation=False, location=False)
meadow_far.data.materials.append(grass_mat)

bpy.context.view_layer.update()

# ==============================================================================
# 8. GEOMETRY NODES WORLD SPACE DISTANCE CULLING NETWORKS (FAILSAFE)
# ==============================================================================
gn_tree = bpy.data.node_groups.new(name="Grass_Scatter_Tree", type='GeometryNodeTree')
gn_nodes = gn_tree.nodes
gn_links = gn_tree.links

gn_tree.interface.new_socket(name="Geometry", in_out='INPUT', socket_type='NodeSocketGeometry')
gn_tree.interface.new_socket(name="Geometry", in_out='OUTPUT', socket_type='NodeSocketGeometry')

group_input = gn_nodes.new('NodeGroupInput')
group_output = gn_nodes.new('NodeGroupOutput')
group_input.location = (-600, 0)
group_output.location = (1000, 0)

cam_info = gn_nodes.new('GeometryNodeObjectInfo')
cam_info.inputs['Object'].default_value = camera
cam_info.location = (-600, 300)

pos_node = gn_nodes.new('GeometryNodeInputPosition')
pos_node.location = (-600, -100)

dist_node = gn_nodes.new('ShaderNodeVectorMath')
dist_node.operation = 'DISTANCE'
dist_node.location = (-300, 0)
gn_links.new(pos_node.outputs['Position'], dist_node.inputs[0])
gn_links.new(cam_info.outputs['Location'], dist_node.inputs[1])

visibility_mask = gn_nodes.new('ShaderNodeMath')
visibility_mask.operation = 'LESS_THAN'
visibility_mask.inputs[1].default_value = 55.0
visibility_mask.location = (0, 0)
gn_links.new(dist_node.outputs['Value'], visibility_mask.inputs[0])

distribute = gn_nodes.new(type="GeometryNodeDistributePointsOnFaces")
distribute.distribute_method = 'POISSON'
distribute.inputs['Distance Min'].default_value = 0.08  # Slighly adjusted density spacing
distribute.inputs['Density Max'].default_value = 140.0  # Reduced slightly for huge BVH build speeds
distribute.location = (300, -100)

gn_links.new(group_input.outputs['Geometry'], distribute.inputs['Mesh'])
gn_links.new(visibility_mask.outputs['Value'], distribute.inputs['Selection'])

instance = gn_nodes.new(type="GeometryNodeInstanceOnPoints")
instance.location = (550, -100)
obj_info = gn_nodes.new(type="GeometryNodeObjectInfo")
obj_info.inputs['Object'].default_value = blade
obj_info.location = (300, -300)

rand_scale = gn_nodes.new(type="FunctionNodeRandomValue")
rand_scale.inputs['Min'].default_value = 0.65
rand_scale.inputs['Max'].default_value = 1.35
rand_scale.location = (300, -500)

rand_tilt_x = gn_nodes.new('FunctionNodeRandomValue')
rand_tilt_x.inputs['Min'].default_value = -0.35
rand_tilt_x.inputs['Max'].default_value = 0.35

rand_tilt_y = gn_nodes.new('FunctionNodeRandomValue')
rand_tilt_y.inputs['Min'].default_value = -0.35
rand_tilt_y.inputs['Max'].default_value = 0.35

rand_spin_z = gn_nodes.new('FunctionNodeRandomValue')
rand_spin_z.inputs['Min'].default_value = 0.0
rand_spin_z.inputs['Max'].default_value = 6.2831

combine_rot = gn_nodes.new('ShaderNodeCombineXYZ')
combine_rot.location = (300, -400)

set_mat = gn_nodes.new(type="GeometryNodeSetMaterial")
set_mat.inputs['Material'].default_value = grass_mat
set_mat.location = (750, -100)

join = gn_nodes.new(type="GeometryNodeJoinGeometry")
join.location = (750, 100)

gn_links.new(distribute.outputs['Points'], instance.inputs['Points'])
gn_links.new(obj_info.outputs['Geometry'], instance.inputs['Instance'])
gn_links.new(rand_tilt_x.outputs['Value'], combine_rot.inputs['X'])
gn_links.new(rand_tilt_y.outputs['Value'], combine_rot.inputs['Y'])
gn_links.new(rand_spin_z.outputs['Value'], combine_rot.inputs['Z'])
gn_links.new(combine_rot.outputs['Vector'], instance.inputs['Rotation'])
gn_links.new(rand_scale.outputs['Value'], instance.inputs['Scale'])
gn_links.new(instance.outputs['Instances'], set_mat.inputs['Geometry'])
gn_links.new(group_input.outputs['Geometry'], join.inputs['Geometry'])
gn_links.new(set_mat.outputs['Geometry'], join.inputs['Geometry'])
gn_links.new(join.outputs['Geometry'], group_output.inputs['Geometry'])

for target_meadow in [meadow_near, meadow_far]:
    geo_modifier = target_meadow.modifiers.new(name="Grass_Scatter_System", type='NODES')
    geo_modifier.node_group = gn_tree

# Scatter Boulders
boulder_configs = [
    (-9.5, 15, 1.0, (4.5, 3.5, 3.8)),
    (-14.0, 28, 2.5, (7.5, 6.0, 7.0)),
    (9.5, 14, 1.0, (4.0, 4.0, 3.5)),
    (14.0, 25, 2.0, (6.0, 5.5, 5.5))
]
boulders = []

for i, (x, y, z, scale_xyz) in enumerate(boulder_configs):
    bpy.ops.mesh.primitive_ico_sphere_add(subdivisions=3, radius=1.0, location=(x, y, z))
    boulder = bpy.context.object
    boulder.name = f"Peripheral_Boulder_{i}"
    boulder.scale = scale_xyz
    boulder.data.materials.append(boulder_mat)
    boulders.append(boulder)

# ==============================================================================
# 9. MONSTER BASE MATERIAL CONFIGURATION
# ==============================================================================
rock_mat = bpy.data.materials.new(name="Rock_Material")
rock_mat.use_nodes = True
r_nodes = rock_mat.node_tree.nodes
r_links = rock_mat.node_tree.links
r_nodes.clear()

node_r_output = r_nodes.new(type='ShaderNodeOutputMaterial')
node_r_principled = r_nodes.new(type='ShaderNodeBsdfPrincipled')
node_r_principled.inputs['Base Color'].default_value = (0.72, 0.28, 0.06, 1.0)
node_r_principled.inputs['Roughness'].default_value = 0.98

node_r_noise = r_nodes.new(type='ShaderNodeTexNoise')
node_r_noise.inputs['Scale'].default_value = 18.0
node_r_noise.inputs['Detail'].default_value = 4.0  # Reduced from 8.0

node_r_bump = r_nodes.new(type='ShaderNodeBump')
node_r_bump.inputs['Strength'].default_value = 0.75

r_links.new(node_r_noise.outputs['Fac'], node_r_bump.inputs['Height'])
r_links.new(node_r_bump.outputs['Normal'], node_r_principled.inputs['Normal'])
r_links.new(node_r_principled.outputs['BSDF'], node_r_output.inputs['Surface'])

def apply_rock_material(obj):
    if len(obj.data.materials) == 0:
        obj.data.materials.append(rock_mat)
    else:
        obj.data.materials[0] = rock_mat

# ==============================================================================
# 10. CHARACTER RIG STRUCTURAL GEOMETRY (THE MONSTER)
# ==============================================================================
monster_parts = []
root_z_height = 2.12

bpy.ops.object.empty_add(type='PLAIN_AXES', location=(6, 0, root_z_height))
root = bpy.context.object
root.name = "Stick_Root"

bpy.ops.mesh.primitive_cylinder_add(radius=0.32, depth=1.2, location=(0, 0, 0))
torso = bpy.context.object
torso.name = "Torso"
torso.parent = root
torso.location = (0, 0, 0)

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=2)
bpy.ops.object.mode_set(mode='OBJECT')
apply_rock_material(torso)
monster_parts.append(torso)

bpy.ops.mesh.primitive_uv_sphere_add(radius=0.34, location=(0, 0, 0.85))
head = bpy.context.object
head.name = "Head"
head.parent = torso
apply_rock_material(head)
monster_parts.append(head)

bpy.ops.mesh.primitive_cylinder_add(radius=0.18, depth=0.9, location=(0, 0, 0.5))
shoulders = bpy.context.object
shoulders.name = "Shoulders"
shoulders.parent = torso
shoulders.rotation_euler = (0, math.radians(90), 0)

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=2)
bpy.ops.object.mode_set(mode='OBJECT')
apply_rock_material(shoulders)
monster_parts.append(shoulders)

bpy.ops.mesh.primitive_cylinder_add(radius=0.15, depth=0.6, location=(0, 0, -0.6))
hips = bpy.context.object
hips.name = "Hips"
hips.parent = torso
hips.rotation_euler = (0, math.radians(90), 0)

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=2)
bpy.ops.object.mode_set(mode='OBJECT')
apply_rock_material(hips)
monster_parts.append(hips)

def create_limb(name, parent, local_position, length=1.0, radius=0.04):
    bpy.ops.mesh.primitive_cylinder_add(radius=radius, depth=length, location=(0, 0, 0))
    limb = bpy.context.object
    limb.name = name
    bpy.ops.object.mode_set(mode='EDIT')
    bpy.ops.mesh.select_all(action='SELECT')
    bpy.ops.mesh.subdivide(number_cuts=1)
    bpy.ops.transform.translate(value=(0, 0, -length / 2))
    bpy.ops.object.mode_set(mode='OBJECT')
    limb.parent = parent
    limb.location = local_position
    apply_rock_material(limb)
    monster_parts.append(limb)
    return limb

def attach_hand(name, parent, local_position):
    bpy.ops.mesh.primitive_uv_sphere_add(radius=0.16, location=(0, 0, 0))
    hand = bpy.context.object
    hand.name = name
    hand.parent = parent
    hand.location = local_position
    apply_rock_material(hand)
    monster_parts.append(hand)
    return hand

def attach_foot(name, parent, local_position):
    bpy.ops.mesh.primitive_cube_add(size=1, location=(0, 0, 0))
    foot = bpy.context.object
    foot.name = name
    bpy.ops.object.mode_set(mode='EDIT')
    bpy.ops.mesh.select_all(action='SELECT')
    bpy.ops.transform.resize(value=(0.34, 0.44, 0.16))
    bpy.ops.transform.translate(value=(0, 0.16, -0.02))
    bpy.ops.object.mode_set(mode='OBJECT')
    foot.parent = parent
    foot.location = local_position
    apply_rock_material(foot)
    monster_parts.append(foot)
    return foot

upper_arm_l = create_limb("Upper_Arm_L", torso, (-0.45, 0, 0.5), length=0.45, radius=0.14)
forearm_l = create_limb("Forearm_L", upper_arm_l, (0, 0, -0.45), length=0.45, radius=0.12)
hand_l = attach_hand("Hand_L", forearm_l, (0, 0, -0.45))

upper_arm_r = create_limb("Upper_Arm_R", torso, (0.45, 0, 0.5), length=0.45, radius=0.14)
forearm_r = create_limb("Forearm_R", upper_arm_r, (0, 0, -0.45), length=0.45, radius=0.12)
hand_r = attach_hand("Hand_R", forearm_r, (0, 0, -0.45))

leg_segment_length = 0.65
thigh_l = create_limb("Thigh_L", torso, (-0.3, 0, -0.6), length=leg_segment_length, radius=0.16)
shin_l = create_limb("Shin_L", thigh_l, (0, 0, -leg_segment_length), length=leg_segment_length, radius=0.13)
foot_l = attach_foot("Foot_L", shin_l, (0, 0, -leg_segment_length))

thigh_r = create_limb("Thigh_R", torso, (0.3, 0, -0.6), length=leg_segment_length, radius=0.16)
shin_r = create_limb("Shin_R", thigh_r, (0, 0, -leg_segment_length), length=leg_segment_length, radius=0.13)
foot_r = attach_foot("Foot_R", shin_r, (0, 0, -leg_segment_length))

# ==============================================================================
# 11. SCROLL ASSET PROP & IMAGE GENERATOR SYSTEM (UPDATED FOR ASPECT RATIO)
# ==============================================================================
bpy.ops.mesh.primitive_cylinder_add(radius=0.025, depth=0.45)
scroll_roll_r = bpy.context.object
scroll_roll_r.name = "Scroll_Roll_R"

bpy.ops.mesh.primitive_cylinder_add(radius=0.025, depth=0.45)
scroll_roll_l = bpy.context.object
scroll_roll_l.name = "Scroll_Roll_L"

# Create a clean baseline plane (1 unit x 1 unit)
bpy.ops.mesh.primitive_plane_add(size=1)
scroll_paper = bpy.context.object
scroll_paper.name = "Scroll_Paper"

# Explicitly clear any existing transformations to ensure scale is 1.0, 1.0, 1.0
scroll_paper.location = (0, 0, 0)
#scroll_paper.rotation_euler = (0, 0, 0)
scroll_paper.scale = (1.0, 1.0, 1.0)

# ... [Your image loading and aspect ratio calculation code here] ...
#aspect_ratio = img_height / img_width # Will correctly be 0.5059

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.transform.rotate(value=math.radians(90), orient_axis='X')
# Reset local scale to baseline 1.0 on X and Y for predictable GN/Matrix math
bpy.ops.transform.resize(value=(1.0, 1.0, 1.0))
bpy.ops.object.mode_set(mode='OBJECT')

mat = bpy.data.materials.new(name="Scroll_Texture_Material")
mat.use_nodes = True
nodes = mat.node_tree.nodes
links = mat.node_tree.links
nodes.clear()

node_tex_coord = nodes.new(type='ShaderNodeTexCoord')
node_mapping = nodes.new(type='ShaderNodeMapping')
node_texture = nodes.new(type='ShaderNodeTexImage')
node_principled = nodes.new(type='ShaderNodeBsdfPrincipled')
node_output = nodes.new(type='ShaderNodeOutputMaterial')

node_mapping.vector_type = 'POINT'
node_texture.interpolation = 'Smart'
image_path = "/Users/dobbinp1/Downloads/IMGP6239_psedit.jpg"

# Baseline fallback ratio (Square)
img_width = 1024
img_height = 1024

if os.path.exists(image_path):
    try:
        img = bpy.data.images.load(image_path)
        # Force Blender to update image properties in memory
        img.gl_load()
        node_texture.image = img
        img_width = img.size[0]
        img_height = img.size[1]
        print(f"Loaded image size: {img_width}x{img_height}")
    except Exception as e:
        print(f"Image load error: {e}")
else:
    img = bpy.data.images.new("Fallback_Grid", width=img_width, height=img_height)
    img.generated_type = 'GRID'
    node_texture.image = img

# --- FIX SECTION 11: REMOVE EDIT MODE MESH ROTATION ---
# Calculate aspect ratio factor (Height / Width)
aspect_ratio = img_height / img_width
print(f"Calculated Aspect Ratio: {aspect_ratio}")

# Keep mesh native coordinates flat (X=Width, Y=Height). Remove the Edit Mode rotation!
bpy.ops.object.mode_set(mode='OBJECT')

# Ensure the texture mapping node is set to clean UV coordinates
node_mapping.vector_type = 'POINT'
node_texture.extension = 'CLIP'

links.new(node_tex_coord.outputs['UV'], node_mapping.inputs['Vector'])
links.new(node_mapping.outputs['Vector'], node_texture.inputs['Vector'])
links.new(node_texture.outputs['Color'], node_principled.inputs['Base Color'])
links.new(node_principled.outputs['BSDF'], node_output.inputs['Surface'])
scroll_paper.data.materials.append(mat)

# ==============================================================================
# 12. INVERSE KINEMATICS MATRIX CALCULATIONS
# ==============================================================================
target_alpha = math.radians(30)
target_gamma = math.radians(30)
eff_length = 0.45 * math.cos(target_alpha) + 0.45 * math.cos(target_alpha + target_gamma)
target_beta_r = math.asin(0.45 / eff_length)
target_beta_l = -target_beta_r



# ==============================================================================
# 13. PATH GENERATOR KEYFRAMES & MAIN SEQUENCE ANIMATION (TWO-PASS FIX)
# ==============================================================================
# Set keyframes for the root object path
scene.frame_set(1)
root.location = (6, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(90))
root.keyframe_insert(data_path="location")
root.keyframe_insert(data_path="rotation_euler")

scene.frame_set(65)
root.location = (0, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(90))
root.keyframe_insert(data_path="location")
root.keyframe_insert(data_path="rotation_euler")

scene.frame_set(75)
root.location = (0, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(180))
root.keyframe_insert(data_path="location")
root.keyframe_insert(data_path="rotation_euler")

for f in [92, 105, 125, 145, 160]:
    scene.frame_set(f)
    root.location = (0, 0, root_z_height)
    root.rotation_euler = (0, 0, math.radians(180))
    root.keyframe_insert(data_path="location")
    root.keyframe_insert(data_path="rotation_euler")


# --- PASS 1: DETERMINING THE ABSOLUTE MAX HAND SPAN ---
absolute_max_span = 0.01

for frame in range(1, 161):
    # Set time context to evaluate limb math positions
    scene.frame_set(frame)

    # Calculate arm angles based on current frame timeline
    if frame < 92:
        alpha = 0
        gamma = 0
    elif 92 <= frame < 105:
        t = (frame - 92) / (105 - 92)
        alpha = target_alpha * t
        gamma = target_gamma * t
    elif 105 <= frame < 125:
        alpha = target_alpha
        gamma = target_gamma
    elif 125 <= frame < 145:
        alpha = target_alpha
        gamma = target_gamma
    else:
        t = (frame - 145) / (160 - 145)
        alpha = target_alpha * (1.0 - t)
        gamma = target_gamma * (1.0 - t)

    upper_arm_r.rotation_euler = (alpha, math.radians(-15), 0)
    forearm_r.rotation_euler = (gamma, 0, 0)
    upper_arm_l.rotation_euler = (alpha, math.radians(15), 0)
    forearm_l.rotation_euler = (gamma, 0, 0)

    # Force dependency graph update to capture actual world space translation
    bpy.context.view_layer.update()

    wpos_r = hand_r.matrix_world.to_translation()
    wpos_l = hand_l.matrix_world.to_translation()

    current_physical_distance = (wpos_l - wpos_r).length
    if current_physical_distance > absolute_max_span:
        absolute_max_span = current_physical_distance

print(f"--> Measured Absolute Maximum Hand Span: {absolute_max_span}")


# --- PASS 2: GENERATE THE TRUE PROP TRANSFORM & KEYFRAMES ---
for frame in range(1, 161):
    scene.frame_set(frame)

    # Recalculate/Reapply rotations so they match the frame evaluation context
    if frame < 92:
        alpha = 0
        gamma = 0
        unfurl_factor = 0.0
    elif 92 <= frame < 105:
        t = (frame - 92) / (105 - 92)
        alpha = target_alpha * t
        gamma = target_gamma * t
        unfurl_factor = 0.0
    elif 105 <= frame < 125:
        t = (frame - 105) / (125 - 105)
        alpha = target_alpha
        gamma = target_gamma
        unfurl_factor = t
    elif 125 <= frame < 145:
        alpha = target_alpha
        gamma = target_gamma
        unfurl_factor = 1.0
    else:
        t = (frame - 145) / (160 - 145)
        alpha = target_alpha * (1.0 - t)
        gamma = target_gamma * (1.0 - t)
        unfurl_factor = 1.0 - t

    # Apply structural limb rotations
    upper_arm_r.rotation_euler = (alpha, math.radians(-15), 0)
    forearm_r.rotation_euler = (gamma, 0, 0)
    upper_arm_l.rotation_euler = (alpha, math.radians(15), 0)
    forearm_l.rotation_euler = (gamma, 0, 0)

    for limb in [thigh_l, thigh_r, shin_l, shin_r, foot_l, foot_r, upper_arm_l, forearm_l, upper_arm_r, forearm_r]:
        limb.keyframe_insert(data_path="rotation_euler")

    # Force immediate dependency refresh
    bpy.context.view_layer.update()

    # Extract calculated positions directly from the matrix translations
    wpos_r = hand_r.matrix_world.to_translation()
    wpos_l = hand_l.matrix_world.to_translation()

    # Position the scroll cylinders relative to the hand reach trajectory
    scroll_roll_r.location = wpos_r
    scroll_roll_l.location = wpos_r + unfurl_factor * (wpos_l - wpos_r)
    scroll_paper.location = (scroll_roll_r.location + scroll_roll_l.location) / 2.0

    # 1. CALCULATE EXPLICIT SCALING ALONG THE DISTANCE VECTOR
    # Multiply the true structural hand-to-hand span by the current unfurl ratio
    current_span = absolute_max_span * unfurl_factor

    target_width = max(0.01, current_span)
    target_height = absolute_max_span * aspect_ratio

    # Assign explicitly to the transform properties
    scroll_paper.scale = (target_width, target_height, 1.0)

    # 2. ROTATION ALIGNMENT
    scroll_roll_r.rotation_euler = (math.radians(90), 0, 0)
    scroll_roll_l.rotation_euler = (math.radians(90), 0, 0)
    scroll_paper.rotation_euler = (math.radians(90), 0, 0)

    # 3. TEXTURE MAPPING STRETCH CONTROL
    u_scale = absolute_max_span / target_width if target_width > 0 else 1.0
    node_mapping.inputs['Scale'].default_value = (u_scale, 1.0, 1.0)
    texture_offset = -0.5 * u_scale + 0.5
    node_mapping.inputs['Location'].default_value = (texture_offset, 0.0, 0.0)

    # Hard-commit scale changes into the keyframe buffer explicitly
    scroll_paper.keyframe_insert(data_path="scale")

    # Commit remaining transforms
    for prop in [scroll_roll_r, scroll_roll_l, scroll_paper]:
        prop.keyframe_insert(data_path="location")
        prop.keyframe_insert(data_path="rotation_euler")

    node_mapping.inputs['Scale'].keyframe_insert(data_path="default_value")
    node_mapping.inputs['Location'].keyframe_insert(data_path="default_value")

print("Optimized Consolidated script executed successfully with two-pass aspect tracking!")
# ==============================================================================
# 14. SURFACE DETAIL INJECTION (OPTIMIZED MODIFIERS)
# ==============================================================================
rock_displace_tex = bpy.data.textures.new(name="Rock_Micro_Deform", type='CLOUDS')
rock_displace_tex.noise_scale = 0.08

for part in monster_parts:
    subsurf = part.modifiers.new(name="Rock_Subsurf", type='SUBSURF')
    subsurf.levels = 1
    subsurf.render_levels = 1

    displace = part.modifiers.new(name="Rock_Displace", type='DISPLACE')
    displace.texture = rock_displace_tex
    displace.strength = 0.085

for b in boulders:
    mesh = b.data
    for poly in mesh.polygons:
        poly.use_smooth = True

if "Camera" in bpy.data.objects:
    bpy.context.scene.camera = bpy.data.objects["Camera"]

view_3d_found = False
for window in bpy.context.window_manager.windows:
    screen = window.screen
    for area in screen.areas:
        if area.type == 'VIEW_3D':
            space = area.spaces.active
            if space and space.type == 'VIEW_3D':
                space.region_3d.view_perspective = 'CAMERA'
                view_3d_found = True
                break
    if view_3d_found:
        break

# Primary Track Tracking Shots
scene.frame_set(1)
camera.location = (0, -12, 2.5)
camera.keyframe_insert(data_path="location")

scene.frame_set(105)
camera.location = (0, -12, 2.5)
camera.keyframe_insert(data_path="location")

scene.frame_set(125)
camera.location = (0, -5.2, 2.0)
camera.keyframe_insert(data_path="location")

scene.frame_set(1)
print("Optimized Consolidated script executed successfully with synchronized scroll tracking!")
```

```
import bpy
import math
import mathutils
import os

# ==============================================================================
# 1. ENVIRONMENT SETUP, SCENE CLEANING & RENDER OPTIMIZATIONS
# ==============================================================================
if bpy.context.object and bpy.context.object.mode != 'OBJECT':
    bpy.ops.object.mode_set(mode='OBJECT')

bpy.ops.object.select_all(action='SELECT')
bpy.ops.object.delete()

scene = bpy.context.scene
scene.frame_start = 1
scene.frame_end = 160
scene.frame_current = 1
scene.render.engine = 'CYCLES'

# --- CYCLES SPEED OPTIMIZATIONS ---
preferences = bpy.context.preferences
addons = preferences.addons
if "cycles" in addons:
    cpref = preferences.addons['cycles'].preferences
    cpref.refresh_devices()
    for device_type in ('OPTIX', 'CUDA', 'OPENCL'):
        try:
            cpref.compute_device_type = device_type
            scene.cycles.device = 'GPU'
            break
        except TypeError:
            pass

scene.cycles.use_adaptive_sampling = True
scene.cycles.adaptive_threshold = 0.05
scene.cycles.samples = 64
scene.cycles.preview_samples = 16

scene.cycles.max_bounces = 6
scene.cycles.diffuse_bounces = 4
scene.cycles.glossy_bounces = 4
scene.cycles.transmission_bounces = 2
scene.cycles.transparent_max_bounces = 4
scene.cycles.volume_bounces = 0

scene.render.use_persistent_data = True

# ==============================================================================
# 2. CAMERA INITIALIZATION
# ==============================================================================
bpy.ops.object.camera_add(location=(0, -12, 2.5), rotation=(math.radians(84), 0, 0))
camera = bpy.context.object
scene.camera = camera

# ==============================================================================
# 3. LIGHT RIG CALIBRATION
# ==============================================================================
bpy.ops.object.select_all(action='DESELECT')
for obj in bpy.data.objects:
    if obj.type == 'LIGHT':
        obj.select_set(True)
bpy.ops.object.delete()

sun_data = bpy.data.lights.new(name="Key_Sun", type='SUN')
sun_data.energy = 2.2
sun_data.angle = 0.15
sun_data.color = (1.0, 0.96, 0.91)
sun_obj = bpy.data.objects.new(name="Key_Sun", object_data=sun_data)
bpy.context.collection.objects.link(sun_obj)
sun_obj.location = (12, -12, 18)
sun_obj.rotation_euler = (0.78, 0.52, 0.65)

fill_data = bpy.data.lights.new(name="Sky_Fill", type='SUN')
fill_data.energy = 0.45
fill_data.color = (0.82, 0.88, 1.0)
fill_obj = bpy.data.objects.new(name="Sky_Fill", object_data=fill_data)
bpy.context.collection.objects.link(fill_obj)
fill_obj.rotation_euler = (-0.6, -0.4, 0.0)

if bpy.context.scene.world:
    world_nodes = bpy.context.scene.world.node_tree.nodes
    bg_node = next((n for n in world_nodes if n.type == 'BACKGROUND'), None)
    if bg_node:
        bg_node.inputs['Strength'].default_value = 0.75
        bg_node.inputs['Color'].default_value = (0.45, 0.55, 0.7, 1.0)

bpy.context.scene.view_settings.view_transform = 'AgX'
bpy.context.scene.view_settings.look = 'AgX - Punchy'
bpy.context.scene.view_settings.exposure = -3.5

# ==============================================================================
# 4. PROCEDURAL SKY MATERIAL SYSTEM
# ==============================================================================
world = scene.world
world.use_nodes = True
wnodes = world.node_tree.nodes
wlinks = world.node_tree.links
wnodes.clear()

node_w_output = wnodes.new(type='ShaderNodeOutputWorld')
node_w_background = wnodes.new(type='ShaderNodeBackground')
node_w_sky = wnodes.new(type='ShaderNodeTexSky')

node_w_sky.sky_type = 'MULTIPLE_SCATTERING'
node_w_sky.sun_elevation = math.radians(16)
node_w_sky.sun_rotation = math.radians(130)
node_w_sky.altitude = 50.0
node_w_sky.air_density = 1.0
node_w_sky.aerosol_density = 0.4
node_w_sky.ozone_density = 2.0

wlinks.new(node_w_sky.outputs['Color'], node_w_background.inputs['Color'])
wlinks.new(node_w_background.outputs['Background'], node_w_output.inputs['Surface'])

# ==============================================================================
# 5. SHADER MATERIAL INITIALIZATION (GRASS & BOULDERS)
# ==============================================================================
grass_mat = bpy.data.materials.new(name="Grass_Material")
grass_mat.use_nodes = True
g_nodes = grass_mat.node_tree.nodes
g_links = grass_mat.node_tree.links
g_nodes.clear()

node_g_output = g_nodes.new(type='ShaderNodeOutputMaterial')
node_g_principled = g_nodes.new(type='ShaderNodeBsdfPrincipled')
node_g_principled.inputs['Base Color'].default_value = (0.12, 0.38, 0.08, 1.0)
node_g_principled.inputs['Roughness'].default_value = 0.6

if 'Subsurface Weight' in node_g_principled.inputs:
    node_g_principled.inputs['Subsurface Weight'].default_value = 0.25
elif 'Subsurface' in node_g_principled.inputs:
    node_g_principled.inputs['Subsurface'].default_value = 0.25

g_links.new(node_g_principled.outputs['BSDF'], node_g_output.inputs['Surface'])

boulder_mat = bpy.data.materials.new(name="Boulder_Material")
boulder_mat.use_nodes = True
b_nodes = boulder_mat.node_tree.nodes
b_links = boulder_mat.node_tree.links
b_nodes.clear()

node_b_output = b_nodes.new(type='ShaderNodeOutputMaterial')
node_b_principled = b_nodes.new(type='ShaderNodeBsdfPrincipled')
node_b_principled.inputs['Roughness'].default_value = 0.95

node_b_coord = b_nodes.new(type='ShaderNodeTexCoord')
node_b_mapping = b_nodes.new(type='ShaderNodeMapping')
node_b_mapping.inputs['Scale'].default_value = (2.0, 2.0, 2.0)
b_links.new(node_b_coord.outputs['Generated'], node_b_mapping.inputs['Vector'])

node_b_color_noise = b_nodes.new(type='ShaderNodeTexNoise')
node_b_color_noise.inputs['Scale'].default_value = 3.0
node_b_color_noise.inputs['Detail'].default_value = 4.0
b_links.new(node_b_mapping.outputs['Vector'], node_b_color_noise.inputs['Vector'])

node_b_ramp = b_nodes.new(type='ShaderNodeValToRGB')
node_b_ramp.color_ramp.elements[0].position = 0.3
node_b_ramp.color_ramp.elements[0].color = (0.28, 0.24, 0.21, 1.0)
node_b_ramp.color_ramp.elements[1].position = 0.7
node_b_ramp.color_ramp.elements[1].color = (0.42, 0.44, 0.45, 1.0)
b_links.new(node_b_color_noise.outputs['Color'], node_b_ramp.inputs['Fac'])
b_links.new(node_b_ramp.outputs['Color'], node_b_principled.inputs['Base Color'])

node_b_bump_noise = b_nodes.new(type='ShaderNodeTexNoise')
node_b_bump_noise.inputs['Scale'].default_value = 24.0
node_b_bump_noise.inputs['Detail'].default_value = 4.0
b_links.new(node_b_mapping.outputs['Vector'], node_b_bump_noise.inputs['Vector'])

node_b_bump = b_nodes.new(type='ShaderNodeBump')
node_b_bump.inputs['Strength'].default_value = 0.45
node_b_bump.inputs['Distance'].default_value = 0.05
b_links.new(node_b_bump_noise.outputs['Fac'], node_b_bump.inputs['Height'])
b_links.new(node_b_bump.outputs['Normal'], node_b_principled.inputs['Normal'])
b_links.new(node_b_principled.outputs['BSDF'], node_b_output.inputs['Surface'])

# ==============================================================================
# 6. PROCEDURAL GRAVEL ROAD SHADER
# ==============================================================================
road_mat = bpy.data.materials.new(name="Gravel_Road_Material")
road_mat.use_nodes = True
nodes = road_mat.node_tree.nodes
links = road_mat.node_tree.links
nodes.clear()

output_node = nodes.new(type='ShaderNodeOutputMaterial')
output_node.location = (800, 300)
principled_node = nodes.new(type='ShaderNodeBsdfPrincipled')
principled_node.location = (500, 300)
links.new(principled_node.outputs['BSDF'], output_node.inputs['Surface'])

tex_coord = nodes.new(type='ShaderNodeTexCoord')
tex_coord.location = (-1200, 300)

mapping = nodes.new(type='ShaderNodeMapping')
mapping.location = (-1000, 300)
mapping.inputs['Scale'].default_value = (600.0, 5.0, 1.0)
links.new(tex_coord.outputs['Generated'], mapping.inputs['Vector'])

soil_noise = nodes.new(type='ShaderNodeTexNoise')
soil_noise.location = (-700, 500)
soil_noise.inputs['Scale'].default_value = 4.0
soil_noise.inputs['Detail'].default_value = 4.0
links.new(mapping.outputs['Vector'], soil_noise.inputs['Vector'])

soil_ramp = nodes.new(type='ShaderNodeValToRGB')
soil_ramp.location = (-450, 500)
soil_ramp.color_ramp.elements[0].color = (0.16, 0.11, 0.08, 1.0)
soil_ramp.color_ramp.elements[1].color = (0.35, 0.28, 0.22, 1.0)
links.new(soil_noise.outputs['Color'], soil_ramp.inputs['Fac'])

gravel_voronoi = nodes.new(type='ShaderNodeTexVoronoi')
gravel_voronoi.location = (-700, 150)
gravel_voronoi.distance = 'EUCLIDEAN'
gravel_voronoi.feature = 'F1'
gravel_voronoi.inputs['Scale'].default_value = 85.0
gravel_voronoi.inputs['Randomness'].default_value = 0.9
links.new(mapping.outputs['Vector'], gravel_voronoi.inputs['Vector'])

stone_ramp = nodes.new(type='ShaderNodeValToRGB')
stone_ramp.location = (-450, 150)
stone_ramp.color_ramp.elements[0].color = (0.32, 0.32, 0.33, 1.0)
stone_ramp.color_ramp.elements[1].color = (0.58, 0.56, 0.54, 1.0)
links.new(gravel_voronoi.outputs['Color'], stone_ramp.inputs['Fac'])

mix_color = nodes.new(type='ShaderNodeMix')
mix_color.data_type = 'RGBA'
mix_color.location = (-150, 350)
mix_color.inputs['Factor'].default_value = 0.65
links.new(soil_ramp.outputs['Color'], mix_color.inputs[6])
links.new(stone_ramp.outputs['Color'], mix_color.inputs[7])
links.new(mix_color.outputs['Result'], principled_node.inputs['Base Color'])

bump_node = nodes.new(type='ShaderNodeBump')
bump_node.location = (150, 50)
bump_node.inputs['Strength'].default_value = 0.90
bump_node.inputs['Distance'].default_value = 0.03
links.new(gravel_voronoi.outputs['Distance'], bump_node.inputs['Height'])
links.new(bump_node.outputs['Normal'], principled_node.inputs['Normal'])

rough_ramp = nodes.new(type='ShaderNodeValToRGB')
rough_ramp.location = (150, 300)
rough_ramp.color_ramp.elements[0].position = 0.2
rough_ramp.color_ramp.elements[0].color = (0.55, 0.55, 0.55, 1.0)
rough_ramp.color_ramp.elements[1].position = 0.8
rough_ramp.color_ramp.elements[1].color = (0.92, 0.92, 0.92, 1.0)
links.new(gravel_voronoi.outputs['Distance'], rough_ramp.inputs['Fac'])
links.new(rough_ramp.outputs['Color'], principled_node.inputs['Roughness'])

road_displace_tex = bpy.data.textures.new(name="Road_Macro_Deform", type='CLOUDS')
road_displace_tex.noise_scale = 0.12

# ==============================================================================
# 7. OPTIMIZED LOW-POLY GRASS CARD ASSET
# ==============================================================================
verts = [
    (-0.04, 0.0, 0.0),
    (0.04, 0.0, 0.0),
    (-0.02, 0.0, 0.25),
    (0.02, 0.0, 0.25),
    (0.0, 0.03, 0.5)
]
faces = [
    (0, 1, 3, 2),
    (2, 3, 4)
]

mesh = bpy.data.meshes.new("Grass_Blade_Mesh")
mesh.from_pydata(verts, [], faces)
mesh.update()

blade = bpy.data.objects.new("Grass_Blade_Asset", mesh)
bpy.context.collection.objects.link(blade)
blade.data.materials.append(grass_mat)

blade.hide_render = True
blade.hide_viewport = True

bpy.ops.mesh.primitive_plane_add(size=1000, location=(0, 0, -0.5))
ground = bpy.context.object
ground.name = "Grass_Terrain_Backdrop"
ground.data.materials.append(grass_mat)

bpy.ops.mesh.primitive_grid_add(x_subdivisions=100, y_subdivisions=10, size=1.0, location=(0, 0, 0.1))
road = bpy.context.object
road.name = "Gravel_Road"
road.scale = (1000.0, 4.0, 1.0)
bpy.ops.object.transform_apply(scale=True, rotation=False, location=False)

road.data.materials.clear()
road.data.materials.append(road_mat)

road_displace = road.modifiers.new(name="Road_3D_Displacement", type='DISPLACE')
road_displace.texture = road_displace_tex
road_displace.strength = 0.08
road_displace.direction = 'Z'

bpy.ops.mesh.primitive_plane_add(size=1, location=(0, -29.0, 0))
meadow_near = bpy.context.object
meadow_near.name = "Grass_Meadow_Near"
meadow_near.scale = (45, 54, 1)
bpy.ops.object.transform_apply(scale=True, rotation=False, location=False)
meadow_near.data.materials.append(grass_mat)

bpy.ops.mesh.primitive_plane_add(size=1, location=(0, 29.0, 0))
meadow_far = bpy.context.object
meadow_far.name = "Grass_Meadow_Far"
meadow_far.scale = (45, 54, 1)
bpy.ops.object.transform_apply(scale=True, rotation=False, location=False)
meadow_far.data.materials.append(grass_mat)

bpy.context.view_layer.update()

# ==============================================================================
# 8. GEOMETRY NODES WORLD SPACE DISTANCE CULLING NETWORKS
# ==============================================================================
gn_tree = bpy.data.node_groups.new(name="Grass_Scatter_Tree", type='GeometryNodeTree')
gn_nodes = gn_tree.nodes
gn_links = gn_tree.links

gn_tree.interface.new_socket(name="Geometry", in_out='INPUT', socket_type='NodeSocketGeometry')
gn_tree.interface.new_socket(name="Geometry", in_out='OUTPUT', socket_type='NodeSocketGeometry')

group_input = gn_nodes.new('NodeGroupInput')
group_output = gn_nodes.new('NodeGroupOutput')
group_input.location = (-600, 0)
group_output.location = (1000, 0)

cam_info = gn_nodes.new('GeometryNodeObjectInfo')
cam_info.inputs['Object'].default_value = camera
cam_info.location = (-600, 300)

pos_node = gn_nodes.new('GeometryNodeInputPosition')
pos_node.location = (-600, -100)

dist_node = gn_nodes.new('ShaderNodeVectorMath')
dist_node.operation = 'DISTANCE'
dist_node.location = (-300, 0)
gn_links.new(pos_node.outputs['Position'], dist_node.inputs[0])
gn_links.new(cam_info.outputs['Location'], dist_node.inputs[1])

visibility_mask = gn_nodes.new('ShaderNodeMath')
visibility_mask.operation = 'LESS_THAN'
visibility_mask.inputs[1].default_value = 55.0
visibility_mask.location = (0, 0)
gn_links.new(dist_node.outputs['Value'], visibility_mask.inputs[0])

distribute = gn_nodes.new(type="GeometryNodeDistributePointsOnFaces")
distribute.distribute_method = 'POISSON'
distribute.inputs['Distance Min'].default_value = 0.08
distribute.inputs['Density Max'].default_value = 140.0
distribute.location = (300, -100)

gn_links.new(group_input.outputs['Geometry'], distribute.inputs['Mesh'])
gn_links.new(visibility_mask.outputs['Value'], distribute.inputs['Selection'])

instance = gn_nodes.new(type="GeometryNodeInstanceOnPoints")
instance.location = (550, -100)
obj_info = gn_nodes.new(type="GeometryNodeObjectInfo")
obj_info.inputs['Object'].default_value = blade
obj_info.location = (300, -300)

rand_scale = gn_nodes.new(type="FunctionNodeRandomValue")
rand_scale.inputs['Min'].default_value = 0.65
rand_scale.inputs['Max'].default_value = 1.35
rand_scale.location = (300, -500)

rand_tilt_x = gn_nodes.new('FunctionNodeRandomValue')
rand_tilt_x.inputs['Min'].default_value = -0.35
rand_tilt_x.inputs['Max'].default_value = 0.35

rand_tilt_y = gn_nodes.new('FunctionNodeRandomValue')
rand_tilt_y.inputs['Min'].default_value = -0.35
rand_tilt_y.inputs['Max'].default_value = 0.35

rand_spin_z = gn_nodes.new('FunctionNodeRandomValue')
rand_spin_z.inputs['Min'].default_value = 0.0
rand_spin_z.inputs['Max'].default_value = 6.2831

combine_rot = gn_nodes.new('ShaderNodeCombineXYZ')
combine_rot.location = (300, -400)

set_mat = gn_nodes.new(type="GeometryNodeSetMaterial")
set_mat.inputs['Material'].default_value = grass_mat
set_mat.location = (750, -100)

join = gn_nodes.new(type="GeometryNodeJoinGeometry")
join.location = (750, 100)

gn_links.new(distribute.outputs['Points'], instance.inputs['Points'])
gn_links.new(obj_info.outputs['Geometry'], instance.inputs['Instance'])
gn_links.new(rand_tilt_x.outputs['Value'], combine_rot.inputs['X'])
gn_links.new(rand_tilt_y.outputs['Value'], combine_rot.inputs['Y'])
gn_links.new(rand_spin_z.outputs['Value'], combine_rot.inputs['Z'])
gn_links.new(combine_rot.outputs['Vector'], instance.inputs['Rotation'])
gn_links.new(rand_scale.outputs['Value'], instance.inputs['Scale'])
gn_links.new(instance.outputs['Instances'], set_mat.inputs['Geometry'])
gn_links.new(group_input.outputs['Geometry'], join.inputs['Geometry'])
gn_links.new(set_mat.outputs['Geometry'], join.inputs['Geometry'])
gn_links.new(join.outputs['Geometry'], group_output.inputs['Geometry'])

for target_meadow in [meadow_near, meadow_far]:
    geo_modifier = target_meadow.modifiers.new(name="Grass_Scatter_System", type='NODES')
    geo_modifier.node_group = gn_tree

boulder_configs = [
    (-9.5, 15, 1.0, (4.5, 3.5, 3.8)),
    (-14.0, 28, 2.5, (7.5, 6.0, 7.0)),
    (9.5, 14, 1.0, (4.0, 4.0, 3.5)),
    (14.0, 25, 2.0, (6.0, 5.5, 5.5))
]
boulders = []

for i, (x, y, z, scale_xyz) in enumerate(boulder_configs):
    bpy.ops.mesh.primitive_ico_sphere_add(subdivisions=3, radius=1.0, location=(x, y, z))
    boulder = bpy.context.object
    boulder.name = f"Peripheral_Boulder_{i}"
    boulder.scale = scale_xyz
    boulder.data.materials.append(boulder_mat)
    boulders.append(boulder)

# ==============================================================================
# 9. MONSTER BASE MATERIAL CONFIGURATION
# ==============================================================================
rock_mat = bpy.data.materials.new(name="Rock_Material")
rock_mat.use_nodes = True
r_nodes = rock_mat.node_tree.nodes
r_links = rock_mat.node_tree.links
r_nodes.clear()

node_r_output = r_nodes.new(type='ShaderNodeOutputMaterial')
node_r_principled = r_nodes.new(type='ShaderNodeBsdfPrincipled')
node_r_principled.inputs['Base Color'].default_value = (0.72, 0.28, 0.06, 1.0)
node_r_principled.inputs['Roughness'].default_value = 0.98

node_r_noise = r_nodes.new(type='ShaderNodeTexNoise')
node_r_noise.inputs['Scale'].default_value = 18.0
node_r_noise.inputs['Detail'].default_value = 4.0

node_r_bump = r_nodes.new(type='ShaderNodeBump')
node_r_bump.inputs['Strength'].default_value = 0.75

r_links.new(node_r_noise.outputs['Fac'], node_r_bump.inputs['Height'])
r_links.new(node_r_bump.outputs['Normal'], node_r_principled.inputs['Normal'])
r_links.new(node_r_principled.outputs['BSDF'], node_r_output.inputs['Surface'])

def apply_rock_material(obj):
    if len(obj.data.materials) == 0:
        obj.data.materials.append(rock_mat)
    else:
        obj.data.materials[0] = rock_mat

# ==============================================================================
# 10. CHARACTER RIG STRUCTURAL GEOMETRY (THE MONSTER)
# ==============================================================================
monster_parts = []
root_z_height = 2.12

bpy.ops.object.empty_add(type='PLAIN_AXES', location=(6, 0, root_z_height))
root = bpy.context.object
root.name = "Stick_Root"

bpy.ops.mesh.primitive_cylinder_add(radius=0.32, depth=1.2, location=(0, 0, 0))
torso = bpy.context.object
torso.name = "Torso"
torso.parent = root
torso.location = (0, 0, 0)

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=2)
bpy.ops.object.mode_set(mode='OBJECT')
apply_rock_material(torso)
monster_parts.append(torso)

bpy.ops.mesh.primitive_uv_sphere_add(radius=0.34, location=(0, 0, 0.85))
head = bpy.context.object
head.name = "Head"
head.parent = torso
apply_rock_material(head)
monster_parts.append(head)

bpy.ops.mesh.primitive_cylinder_add(radius=0.18, depth=0.9, location=(0, 0, 0.5))
shoulders = bpy.context.object
shoulders.name = "Shoulders"
shoulders.parent = torso
shoulders.rotation_euler = (0, math.radians(90), 0)

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=2)
bpy.ops.object.mode_set(mode='OBJECT')
apply_rock_material(shoulders)
monster_parts.append(shoulders)

bpy.ops.mesh.primitive_cylinder_add(radius=0.15, depth=0.6, location=(0, 0, -0.6))
hips = bpy.context.object
hips.name = "Hips"
hips.parent = torso
hips.rotation_euler = (0, math.radians(90), 0)

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=2)
bpy.ops.object.mode_set(mode='OBJECT')
apply_rock_material(hips)
monster_parts.append(hips)

def create_limb(name, parent, local_position, length=1.0, radius=0.04):
    bpy.ops.mesh.primitive_cylinder_add(radius=radius, depth=length, location=(0, 0, 0))
    limb = bpy.context.object
    limb.name = name
    bpy.ops.object.mode_set(mode='EDIT')
    bpy.ops.mesh.select_all(action='SELECT')
    bpy.ops.mesh.subdivide(number_cuts=1)
    bpy.ops.transform.translate(value=(0, 0, -length / 2))
    bpy.ops.object.mode_set(mode='OBJECT')
    limb.parent = parent
    limb.location = local_position
    apply_rock_material(limb)
    monster_parts.append(limb)
    return limb

def attach_hand(name, parent, local_position):
    bpy.ops.mesh.primitive_uv_sphere_add(radius=0.16, location=(0, 0, 0))
    hand = bpy.context.object
    hand.name = name
    hand.parent = parent
    hand.location = local_position
    apply_rock_material(hand)
    monster_parts.append(hand)
    return hand

def attach_foot(name, parent, local_position):
    bpy.ops.mesh.primitive_cube_add(size=1, location=(0, 0, 0))
    foot = bpy.context.object
    foot.name = name
    bpy.ops.object.mode_set(mode='EDIT')
    bpy.ops.mesh.select_all(action='SELECT')
    bpy.ops.transform.resize(value=(0.34, 0.44, 0.16))
    bpy.ops.transform.translate(value=(0, 0.16, -0.02))
    bpy.ops.object.mode_set(mode='OBJECT')
    foot.parent = parent
    foot.location = local_position
    apply_rock_material(foot)
    monster_parts.append(foot)
    return foot

upper_arm_l = create_limb("Upper_Arm_L", torso, (-0.45, 0, 0.5), length=0.45, radius=0.14)
forearm_l = create_limb("Forearm_L", upper_arm_l, (0, 0, -0.45), length=0.45, radius=0.12)
hand_l = attach_hand("Hand_L", forearm_l, (0, 0, -0.45))

upper_arm_r = create_limb("Upper_Arm_R", torso, (0.45, 0, 0.5), length=0.45, radius=0.14)
forearm_r = create_limb("Forearm_R", upper_arm_r, (0, 0, -0.45), length=0.45, radius=0.12)
hand_r = attach_hand("Hand_R", forearm_r, (0, 0, -0.45))

leg_segment_length = 0.65
thigh_l = create_limb("Thigh_L", torso, (-0.3, 0, -0.6), length=leg_segment_length, radius=0.16)
shin_l = create_limb("Shin_L", thigh_l, (0, 0, -leg_segment_length), length=leg_segment_length, radius=0.13)
foot_l = attach_foot("Foot_L", shin_l, (0, 0, -leg_segment_length))

thigh_r = create_limb("Thigh_R", torso, (0.3, 0, -0.6), length=leg_segment_length, radius=0.16)
shin_r = create_limb("Shin_R", thigh_r, (0, 0, -leg_segment_length), length=leg_segment_length, radius=0.13)
foot_r = attach_foot("Foot_R", shin_r, (0, 0, -leg_segment_length))

# ==============================================================================
# 11. SCROLL PROP INITIALIZATION & EXPLICIT ASPECT RATIO MAPPING
# ==============================================================================
image_path = "/Users/dobbinp1/Downloads/IMGP6239_psedit.jpg"
img_width = 1024
img_height = 1024

mat = bpy.data.materials.new(name="Scroll_Texture_Material")
mat.use_nodes = True
nodes = mat.node_tree.nodes
links = mat.node_tree.links
nodes.clear()

node_tex_coord = nodes.new(type='ShaderNodeTexCoord')
node_mapping = nodes.new(type='ShaderNodeMapping')
node_texture = nodes.new(type='ShaderNodeTexImage')
node_principled = nodes.new(type='ShaderNodeBsdfPrincipled')
node_output = nodes.new(type='ShaderNodeOutputMaterial')

node_mapping.vector_type = 'POINT'
node_texture.interpolation = 'Smart'
node_texture.extension = 'CLIP'

if os.path.exists(image_path):
    try:
        img = bpy.data.images.load(image_path)
        img.gl_load()
        node_texture.image = img
        img_width = img.size[0]
        img_height = img.size[1]
        print(f"Loaded image size: {img_width}x{img_height}")
    except Exception as e:
        print(f"Image load error: {e}")
else:
    img = bpy.data.images.new("Fallback_Grid", width=img_width, height=img_height)
    img.generated_type = 'GRID'
    node_texture.image = img

aspect_ratio = img_height / img_width
print(f"Calculated Aspect Ratio: {aspect_ratio}")

# Re-engineered clean baseline: create standard native mesh geometry
# Scroll base width matches the physical texture dimension aspect logic
bpy.ops.mesh.primitive_cylinder_add(radius=0.025, depth=1.0)
scroll_roll_r = bpy.context.object
scroll_roll_r.name = "Scroll_Roll_R"

bpy.ops.mesh.primitive_cylinder_add(radius=0.025, depth=1.0)
scroll_roll_l = bpy.context.object
scroll_roll_l.name = "Scroll_Roll_L"

bpy.ops.mesh.primitive_plane_add(size=1)
scroll_paper = bpy.context.object
scroll_paper.name = "Scroll_Paper"

scroll_paper.location = (0, 0, 0)
scroll_paper.scale = (1.0, 1.0, 1.0)

links.new(node_tex_coord.outputs['UV'], node_mapping.inputs['Vector'])
links.new(node_mapping.outputs['Vector'], node_texture.inputs['Vector'])
links.new(node_texture.outputs['Color'], node_principled.inputs['Base Color'])
links.new(node_principled.outputs['BSDF'], node_output.inputs['Surface'])
scroll_paper.data.materials.append(mat)

# ==============================================================================
# 12. INVERSE KINEMATICS MATRIX CALCULATIONS
# ==============================================================================
target_alpha = math.radians(30)
target_gamma = math.radians(30)
eff_length = 0.45 * math.cos(target_alpha) + 0.45 * math.cos(target_alpha + target_gamma)
target_beta_r = math.asin(0.45 / eff_length)
target_beta_l = -target_beta_r

# ==============================================================================
# 13. MAIN SEQUENCE ANIMATION (STABLE TRANSFORM EVALUATION)
# ==============================================================================
scene.frame_set(1)
root.location = (6, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(90))
root.keyframe_insert(data_path="location")
root.keyframe_insert(data_path="rotation_euler")

scene.frame_set(65)
root.location = (0, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(90))
root.keyframe_insert(data_path="location")
root.keyframe_insert(data_path="rotation_euler")

scene.frame_set(75)
root.location = (0, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(180))
root.keyframe_insert(data_path="location")
root.keyframe_insert(data_path="rotation_euler")

for f in [92, 105, 125, 145, 160]:
    scene.frame_set(f)
    root.location = (0, 0, root_z_height)
    root.rotation_euler = (0, 0, math.radians(180))
    root.keyframe_insert(data_path="location")
    root.keyframe_insert(data_path="rotation_euler")

# PASS 1: Calculate the maximum physical width separation of the hands
absolute_max_span = 0.01

for frame in range(1, 161):
    scene.frame_set(frame)
    if frame < 92:
        alpha = 0
        gamma = 0
    elif 92 <= frame < 105:
        t = (frame - 92) / (105 - 92)
        alpha = target_alpha * t
        gamma = target_gamma * t
    elif 105 <= frame < 125:
        alpha = target_alpha
        gamma = target_gamma
    elif 125 <= frame < 145:
        alpha = target_alpha
        gamma = target_gamma
    else:
        t = (frame - 145) / (160 - 145)
        alpha = target_alpha * (1.0 - t)
        gamma = target_gamma * (1.0 - t)

    upper_arm_r.rotation_euler = (alpha, math.radians(-15), 0)
    forearm_r.rotation_euler = (gamma, 0, 0)
    upper_arm_l.rotation_euler = (alpha, math.radians(15), 0)
    forearm_l.rotation_euler = (gamma, 0, 0)

    bpy.context.view_layer.update()
    wpos_r = hand_r.matrix_world.to_translation()
    wpos_l = hand_l.matrix_world.to_translation()

    current_physical_distance = (wpos_l - wpos_r).length
    if current_physical_distance > absolute_max_span:
        absolute_max_span = current_physical_distance

print(f"--> Measured Absolute Maximum Hand Span: {absolute_max_span}")

# PASS 2: Explicitly drive the physical proportions using the aspect ratio rules
for frame in range(1, 161):
    scene.frame_set(frame)
    if frame < 92:
        alpha = 0
        gamma = 0
        beta_r = 0
        beta_l = 0
        unfurl_factor = 0.0
    elif 92 <= frame < 105:
        t = (frame - 92) / (105 - 92)
        alpha = target_alpha * t
        gamma = target_gamma * t
        beta_r = target_beta_r * t
        beta_l = target_beta_l * t
        unfurl_factor = 0.0
    elif 105 <= frame < 125:
        t = (frame - 105) / (125 - 105)
        alpha = target_alpha
        gamma = target_gamma
        beta_r = target_beta_r + (0 - target_beta_r) * t
        beta_l = target_beta_l + (0 - target_beta_l) * t
        unfurl_factor = t
    elif 125 <= frame < 145:
        alpha = target_alpha
        gamma = target_gamma
        beta_r = 0
        beta_l = 0
        unfurl_factor = 1.0
    else:
        t = (frame - 145) / (160 - 145)
        alpha = target_alpha * (1.0 - t)
        gamma = target_gamma * (1.0 - t)
        beta_r = target_beta_r * t
        beta_l = target_beta_l * t
        unfurl_factor = 1.0 - t

    # Apply synchronized structural limb rotations including hand spatial tracking
    upper_arm_r.rotation_euler = (alpha, math.radians(-15) + beta_r, 0)
    forearm_r.rotation_euler = (gamma, 0, 0)
    upper_arm_l.rotation_euler = (alpha, math.radians(15) + beta_l, 0)
    forearm_l.rotation_euler = (gamma, 0, 0)

    for limb in [thigh_l, thigh_r, shin_l, shin_r, foot_l, foot_r, upper_arm_l, forearm_l, upper_arm_r, forearm_r]:
        limb.keyframe_insert(data_path="rotation_euler")

    bpy.context.view_layer.update()
    wpos_r = hand_r.matrix_world.to_translation()
    wpos_l = hand_l.matrix_world.to_translation()

    # Track locations cleanly based on hand movement paths
    scroll_roll_r.location = wpos_r
    scroll_roll_l.location = wpos_r + unfurl_factor * (wpos_l - wpos_r)
    scroll_paper.location = (scroll_roll_r.location + scroll_roll_l.location) / 2.0

    # Explicit scale tracking: the absolute height matches the max width scaled by aspect ratio
    target_width = max(0.01, absolute_max_span * unfurl_factor)
    target_height = absolute_max_span * aspect_ratio

    # Physical Mesh Object Transform scaling rules
    scroll_paper.scale = (target_width, target_height, 1.0)
    scroll_roll_r.scale = (1.0, 1.0, target_height)
    scroll_roll_l.scale = (1.0, 1.0, target_height)

    scroll_roll_r.rotation_euler = (math.radians(90), 0, 0)
    scroll_roll_l.rotation_euler = (math.radians(90), 0, 0)
    scroll_paper.rotation_euler = (math.radians(90), 0, 0)

    # Dynamic mapping shifts inside the node graph to prevent texture crushing
    u_scale = absolute_max_span / target_width if target_width > 0 else 1.0
    node_mapping.inputs['Scale'].default_value = (u_scale, 1.0, 1.0)
    texture_offset = -0.5 * u_scale + 0.5
    node_mapping.inputs['Location'].default_value = (texture_offset, 0.0, 0.0)

    # Keyframe commits
    scroll_paper.keyframe_insert(data_path="scale")
    scroll_roll_r.keyframe_insert(data_path="scale")
    scroll_roll_l.keyframe_insert(data_path="scale")

    for prop in [scroll_roll_r, scroll_roll_l, scroll_paper]:
        prop.keyframe_insert(data_path="location")
        prop.keyframe_insert(data_path="rotation_euler")

    node_mapping.inputs['Scale'].keyframe_insert(data_path="default_value")
    node_mapping.inputs['Location'].keyframe_insert(data_path="default_value")

# ==============================================================================
# 14. SURFACE DETAIL INJECTION (OPTIMIZED MODIFIERS)
# ==============================================================================
rock_displace_tex = bpy.data.textures.new(name="Rock_Micro_Deform", type='CLOUDS')
rock_displace_tex.noise_scale = 0.08

for part in monster_parts:
    subsurf = part.modifiers.new(name="Rock_Subsurf", type='SUBSURF')
    subsurf.levels = 1
    subsurf.render_levels = 1

    displace = part.modifiers.new(name="Rock_Displace", type='DISPLACE')
    displace.texture = rock_displace_tex
    displace.strength = 0.085

for b in boulders:
    mesh = b.data
    for poly in mesh.polygons:
        poly.use_smooth = True

if "Camera" in bpy.data.objects:
    bpy.context.scene.camera = bpy.data.objects["Camera"]

view_3d_found = False
for window in bpy.context.window_manager.windows:
    screen = window.screen
    for area in screen.areas:
        if area.type == 'VIEW_3D':
            space = area.spaces.active
            if space and space.type == 'VIEW_3D':
                space.region_3d.view_perspective = 'CAMERA'
                view_3d_found = True
                break
    if view_3d_found:
        break

scene.frame_set(1)
camera.location = (0, -12, 2.5)
camera.keyframe_insert(data_path="location")

scene.frame_set(105)
camera.location = (0, -12, 2.5)
camera.keyframe_insert(data_path="location")

scene.frame_set(125)
camera.location = (0, -5.2, 2.0)
camera.keyframe_insert(data_path="location")

scene.frame_set(1)
print("Updated scroll script with dynamic aspect ratio constraints executed successfully!")
```

```
import os
import subprocess
import torch
import shutil


def isolate_voice_with_ai(input_file, output_dir="ai_output"):
    """Isolates vocals/voices from an audio file using the HTDemucs AI model."""

    if not os.path.exists(input_file):
        print(f"Error: Input file '{input_file}' not found.")
        return

    # Check if a GPU is available to speed up the AI processing
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"Using device: {device.upper()} for AI processing.")
    print("Initializing Demucs (this may take a minute on the first run to download the model)...")

    # Command to run Demucs
    # htdemucs is the high-quality Hybrid Transformer model
    # --two-stems=vocals splits it cleanly into 'vocals' and 'no_vocals'
    # Command to run Demucs directly via Python
    command = [
        "/Volumes/macstorage/Users/dobbinp1/Applications/Blender.app/Contents/Resources/5.1/python/bin/python3.13", "-m", "demucs",
        "--two-stems", "vocals",
        "-n", "htdemucs",
        "-d", device,
        "-o", output_dir,
        input_file
    ]


    try:
        # Run the AI separation process
        subprocess.run(command, check=True)

        # Locating the generated vocals file
        # Demucs creates a folder structure: output_dir/model_name/filename/vocals.wav
        filename_only = os.path.splitext(os.path.basename(input_file))[0]
        vocals_path = os.path.join(output_dir, "htdemucs", filename_only, "vocals.wav")

        if os.path.exists(vocals_path):
            final_destination = f"isolated_voice_{filename_only}.wav"
            shutil.copy(vocals_path, final_destination)
            print("\n" + "="*40)
            print("SUCCESS!")
            print(f"The isolated AI voice track has been saved as: {final_destination}")
            print("="*40)
        else:
            print("Process finished, but output file could not be located.")

    except subprocess.CalledProcessError as e:
        print(f"An error occurred while running the AI model: {e}")
    except FileNotFoundError:
        print("Error: 'demucs' command not found. Please ensure you ran 'pip install demucs'.")

if __name__ == "__main__":
    # Update this with your actual m4a filename
    INPUT_M4A_FILE = "/Users/dobbinp1/Downloads/New Recording.m4a"

    isolate_voice_with_ai(INPUT_M4A_FILE)
```
