```
import bpy
import math
import mathutils
import os

# ==========================================
# 1. SETUP ENVIRONMENT & CLEAN SCENE
# ==========================================
if bpy.context.object and bpy.context.object.mode != 'OBJECT':
    bpy.ops.object.mode_set(mode='OBJECT')

bpy.ops.object.select_all(action='SELECT')
bpy.ops.object.delete()

scene = bpy.context.scene
scene.frame_start = 1
scene.frame_end = 160
scene.frame_current = 1

scene.render.engine = 'CYCLES'

# Add standard Camera
bpy.ops.object.camera_add(location=(0, -12, 2.5), rotation=(math.radians(84), 0, 0))
camera = bpy.context.object
scene.camera = camera

# Add a Light source
bpy.ops.object.light_add(type='SUN', location=(5, -10, 10))
sun = bpy.context.object
sun.data.energy = 2.5
sun.rotation_euler = (math.radians(45), math.radians(0), math.radians(35))

# ==========================================
# 1b. PROCEDURAL SKY SETUP
# ==========================================
world = scene.world
world.use_nodes = True
wnodes = world.node_tree.nodes
wlinks = world.node_tree.links
wnodes.clear()

node_w_output = wnodes.new(type='ShaderNodeOutputWorld')
node_w_background = wnodes.new(type='ShaderNodeBackground')
node_w_sky = wnodes.new(type='ShaderNodeTexSky')

node_w_sky.sky_type = 'MULTIPLE_SCATTERING'
node_w_sky.sun_elevation = math.radians(16)
node_w_sky.sun_rotation = math.radians(130)
node_w_sky.altitude = 50.0
node_w_sky.air_density = 1.0
node_w_sky.aerosol_density = 0.4
node_w_sky.ozone_density = 2.0

wlinks.new(node_w_sky.outputs['Color'], node_w_background.inputs['Color'])
wlinks.new(node_w_background.outputs['Background'], node_w_output.inputs['Surface'])

# ==========================================
# 2a. MATERIAL INITIALIZATION
# ==========================================
# Organic Grass Material
grass_mat = bpy.data.materials.new(name="Grass_Material")
grass_mat.use_nodes = True
g_nodes = grass_mat.node_tree.nodes
g_links = grass_mat.node_tree.links
g_nodes.clear()

node_g_output = g_nodes.new(type='ShaderNodeOutputMaterial')
node_g_principled = g_nodes.new(type='ShaderNodeBsdfPrincipled')
node_g_principled.inputs['Base Color'].default_value = (0.12, 0.38, 0.08, 1.0)
node_g_principled.inputs['Roughness'].default_value = 0.6

if 'Subsurface Weight' in node_g_principled.inputs:
    node_g_principled.inputs['Subsurface Weight'].default_value = 0.25
elif 'Subsurface' in node_g_principled.inputs:
    node_g_principled.inputs['Subsurface'].default_value = 0.25

g_links.new(node_g_principled.outputs['BSDF'], node_g_output.inputs['Surface'])


# Dirt Road Material
road_mat = bpy.data.materials.new(name="Road_Material")
# Execute the material update
#update_road_to_gravel("Road_Material")

#road_mat.use_nodes = True
#ro_nodes = road_mat.node_tree.nodes
#ro_links = road_mat.node_tree.links
#ro_nodes.clear()




#node_ro_output = ro_nodes.new(type='ShaderNodeOutputMaterial')
#node_ro_principled = ro_nodes.new(type='ShaderNodeBsdfPrincipled')
#node_ro_principled.inputs['Base Color'].default_value = (0.42, 0.34, 0.26, 1.0)
#node_ro_principled.inputs['Roughness'].default_value = 0.95
#node_ro_noise = ro_nodes.new(type='ShaderNodeTexNoise')
#node_ro_noise.inputs['Scale'].default_value = 40.0
#node_ro_bump = ro_nodes.new(type='ShaderNodeBump')
#node_ro_bump.inputs['Strength'].default_value = 0.3
#ro_links.new(node_ro_noise.outputs['Fac'], node_ro_bump.inputs['Height'])
#ro_links.new(node_ro_bump.outputs['Normal'], node_ro_principled.inputs['Normal'])
#ro_links.new(node_ro_principled.outputs['BSDF'], node_ro_output.inputs['Surface'])

# Find or create the material
#mat = bpy.data.materials.get(material_name)
#if not mat:
#    mat = bpy.data.materials.new(name=material_name)

road_mat.use_nodes = True
nodes = road_mat.node_tree.nodes
links = road_mat.node_tree.links

# Clear existing nodes to rebuild the tree cleanly
nodes.clear()

# 1. Create Core Nodes
output_node = nodes.new(type='ShaderNodeOutputMaterial')
output_node.location = (600, 300)

principled_node = nodes.new(type='ShaderNodeBsdfPrincipled')
principled_node.location = (300, 300)
links.new(principled_node.outputs['BSDF'], output_node.inputs['Surface'])

# Texture Coordinate and Mapping for global scale control
tex_coord = nodes.new(type='ShaderNodeTexCoord')
tex_coord.location = (-1200, 300)

mapping = nodes.new(type='ShaderNodeMapping')
mapping.location = (-1000, 300)
mapping.inputs['Scale'].default_value = (500.0, 2.0, 1.0) # Scale up for dense pebbles
links.new(tex_coord.outputs['UV'], mapping.inputs['Vector'])

# 2. Base Gravel Color & Tire Ruts Generation
# Noise texture to simulate variations in dirt/dust color
dirt_noise = nodes.new(type='ShaderNodeTexNoise')
dirt_noise.location = (-700, 500)
dirt_noise.inputs['Scale'].default_value = 5.0
dirt_noise.inputs['Detail'].default_value = 6.0
links.new(mapping.outputs['Vector'], dirt_noise.inputs['Vector'])

# FIXED: Using 'ShaderNodeValToRGB' instead of 'ShaderNodeColorRamp'
color_ramp = nodes.new(type='ShaderNodeValToRGB')
color_ramp.location = (-400, 500)
# Brownish dirt base to grey gravel rocks
color_ramp.color_ramp.elements[0].color = (0.25, 0.18, 0.13, 1.0) # Dark Dirt
color_ramp.color_ramp.elements[1].color = (0.45, 0.42, 0.40, 1.0) # Grey Stone
links.new(dirt_noise.outputs['Color'], color_ramp.inputs['Fac'])
links.new(color_ramp.outputs['Color'], principled_node.inputs['Base Color'])

# 3. Micro-Pebble Displacement (The Gravel Texture)
pebble_voronoi = nodes.new(type='ShaderNodeTexVoronoi')
pebble_voronoi.location = (-700, 100)
pebble_voronoi.distance = 'EUCLIDEAN'
pebble_voronoi.feature = 'F1'
pebble_voronoi.inputs['Scale'].default_value = 45.0
pebble_voronoi.inputs['Randomness'].default_value = 0.9
links.new(mapping.outputs['Vector'], pebble_voronoi.inputs['Vector'])

# High frequency noise to break up the clean edges of the Voronoi cells
pebble_detail = nodes.new(type='ShaderNodeTexNoise')
pebble_detail.location = (-700, -200)
pebble_detail.inputs['Scale'].default_value = 80.0
pebble_detail.inputs['Detail'].default_value = 4.0
links.new(mapping.outputs['Vector'], pebble_detail.inputs['Vector'])

# Combine the Voronoi shapes and fine noise for the bump map
mix_bump = nodes.new(type='ShaderNodeMix')
mix_bump.data_type = 'FLOAT'
mix_bump.location = (-400, -50)
mix_bump.inputs['Factor'].default_value = 0.35
links.new(pebble_voronoi.outputs['Distance'], mix_bump.inputs[4])
links.new(pebble_detail.outputs['Fac'], mix_bump.inputs[5])

bump_node = nodes.new(type='ShaderNodeBump')
bump_node.location = (-100, -50)
bump_node.inputs['Strength'].default_value = 0.65  # Adjust for surface roughness depth
bump_node.inputs['Distance'].default_value = 0.1
links.new(mix_bump.outputs[0], bump_node.inputs['Height'])
links.new(bump_node.outputs['Normal'], principled_node.inputs['Normal'])

# 4. Surface Roughness
# FIXED: Using 'ShaderNodeValToRGB' here as well
roughness_ramp = nodes.new(type='ShaderNodeValToRGB')
roughness_ramp.location = (-100, 250)
roughness_ramp.color_ramp.elements[0].position = 0.2
roughness_ramp.color_ramp.elements[0].color = (0.5, 0.5, 0.5, 1.0) # Rocks are slightly glossy
roughness_ramp.color_ramp.elements[1].position = 0.8
roughness_ramp.color_ramp.elements[1].color = (0.9, 0.9, 0.9, 1.0) # Dirt matrix is matte
links.new(mix_bump.outputs[0], roughness_ramp.inputs['Fac'])
links.new(roughness_ramp.outputs['Color'], principled_node.inputs['Roughness'])








# --- ADD THIS TO APPLY IT TO THE ROAD OBJECT ---
#road_obj = bpy.data.objects.get("Dirt_Road")
#if road_obj:
#    new_mat = bpy.data.materials.get("RoadMaterial")
#    if new_mat:
#        # Clear existing materials and apply the new gravel material
#        road_obj.data.materials.clear()
#        road_obj.data.materials.append(new_mat)



# Border Boulder Material
boulder_mat = bpy.data.materials.new(name="Boulder_Material")
boulder_mat.use_nodes = True
b_nodes = boulder_mat.node_tree.nodes
b_links = boulder_mat.node_tree.links
b_nodes.clear()

node_b_output = b_nodes.new(type='ShaderNodeOutputMaterial')
node_b_principled = b_nodes.new(type='ShaderNodeBsdfPrincipled')
node_b_principled.inputs['Base Color'].default_value = (0.28, 0.28, 0.28, 1.0)
node_b_principled.inputs['Roughness'].default_value = 0.9
node_b_noise = b_nodes.new(type='ShaderNodeTexNoise')
node_b_noise.inputs['Scale'].default_value = 12.0
node_b_bump = b_nodes.new(type='ShaderNodeBump')
node_b_bump.inputs['Strength'].default_value = 0.6
b_links.new(node_b_noise.outputs['Fac'], node_b_bump.inputs['Height'])
b_links.new(node_b_bump.outputs['Normal'], node_b_principled.inputs['Normal'])
b_links.new(node_b_principled.outputs['BSDF'], node_b_output.inputs['Surface'])

displace_tex = bpy.data.textures.new(name="Rock_Displace_Tex", type='CLOUDS')
displace_tex.noise_scale = 0.2

boulder_displace_tex = bpy.data.textures.new(name="Boulder_Displace_Tex", type='VORONOI')
boulder_displace_tex.noise_scale = 0.6

# ==========================================
# 2b. GENERATE SEGMENTED GEOMETRY
# ==========================================
# 1. Master Grass Blade Object
bpy.ops.mesh.primitive_cone_add(vertices=3, radius1=0.02, depth=0.35, location=(0, 0, -50))
blade = bpy.context.object
blade.name = "Grass_Blade_Asset"
blade.data.materials.append(grass_mat)

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.transform.translate(value=(0, 0, 0.175))
bpy.ops.object.mode_set(mode='OBJECT')
bpy.context.scene.cursor.location = (0, 0, -50)
bpy.ops.object.origin_set(type='ORIGIN_CURSOR')

# 2. Vast Background Grass Terrain
bpy.ops.mesh.primitive_plane_add(size=1000, location=(0, 0, -0.01))
ground = bpy.context.object
ground.name = "Grass_Terrain_Backdrop"
ground.data.materials.append(grass_mat)

# 3. The Dirt Road (Solid, thick block structure center at Y=0)
bpy.ops.mesh.primitive_cube_add(size=1, location=(0, 0, 0))
road = bpy.context.object
road.name = "Dirt_Road"
road.scale = (1000, 4.0, 0.1)
bpy.ops.object.transform_apply(scale=True, rotation=False, location=False)

# FIX: Unwrap the cube so the material maps correctly to the new scale
bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.uv.cube_project() # Or bpy.ops.uv.smart_project()
bpy.ops.object.mode_set(mode='OBJECT')

# Clear any accidental default materials
road.data.materials.clear()
# Append your material to the first slot
road.data.materials.append(road_mat)



# 4. Near Grass Piece (Foreground, in front of the road)
# Road width is 4.0 (stretching from Y = -2.0 to Y = 2.0).
# Near piece covers from Y = -2.0 extending out down the screen.
bpy.ops.mesh.primitive_plane_add(size=1, location=(0, -29.0, 0))
meadow_near = bpy.context.object
meadow_near.name = "Grass_Meadow_Near"
meadow_near.scale = (45, 54, 1) # Extends cleanly right up to the road edge
bpy.ops.object.transform_apply(scale=True, rotation=False, location=False)
meadow_near.data.materials.append(grass_mat)

# 5. Far Grass Piece (Background, behind the road)
# Covers from Y = 2.0 extending backward into the horizon.
bpy.ops.mesh.primitive_plane_add(size=1, location=(0, 29.0, 0))
meadow_far = bpy.context.object
meadow_far.name = "Grass_Meadow_Far"
meadow_far.scale = (45, 54, 1)
bpy.ops.object.transform_apply(scale=True, rotation=False, location=False)
meadow_far.data.materials.append(grass_mat)

# Force a scene update
bpy.context.view_layer.update()

# ==========================================
# 2c. BUILD THE GEOMETRY NODES SCATTERING TREE
# ==========================================
gn_tree = bpy.data.node_groups.new(name="Grass_Scatter_Tree", type='GeometryNodeTree')
gn_nodes = gn_tree.nodes
gn_links = gn_tree.links

gn_tree.interface.new_socket(name="Geometry", in_out='INPUT', socket_type='NodeSocketGeometry')
gn_tree.interface.new_socket(name="Geometry", in_out='OUTPUT', socket_type='NodeSocketGeometry')

group_input = gn_nodes.new('NodeGroupInput')
group_output = gn_nodes.new('NodeGroupOutput')
group_input.location = (-400, 0)
group_output.location = (800, 0)

distribute = gn_nodes.new(type="GeometryNodeDistributePointsOnFaces")
distribute.inputs['Density'].default_value = 85.0

instance = gn_nodes.new(type="GeometryNodeInstanceOnPoints")

obj_info = gn_nodes.new(type="GeometryNodeObjectInfo")
obj_info.inputs['Object'].default_value = blade

rand_scale = gn_nodes.new(type="FunctionNodeRandomValue")
rand_scale.data_type = 'FLOAT'
rand_scale.inputs['Min'].default_value = 0.6
rand_scale.inputs['Max'].default_value = 1.4

rand_rot = gn_nodes.new(type="FunctionNodeRandomValue")
rand_rot.data_type = 'FLOAT_VECTOR'
rand_rot.inputs['Min'].default_value = (-0.15, -0.15, 0.0)
rand_rot.inputs['Max'].default_value = (0.15, 0.15, 6.28)

set_mat = gn_nodes.new(type="GeometryNodeSetMaterial")
set_mat.inputs['Material'].default_value = grass_mat

join = gn_nodes.new(type="GeometryNodeJoinGeometry")

# Connect Main Scatter Logic (No proximity/coordinate mask nodes needed now!)
gn_links.new(group_input.outputs['Geometry'], distribute.inputs['Mesh'])
gn_links.new(distribute.outputs['Points'], instance.inputs['Points'])
gn_links.new(obj_info.outputs['Geometry'], instance.inputs['Instance'])
gn_links.new(rand_scale.outputs['Value'], instance.inputs['Scale'])
gn_links.new(rand_rot.outputs['Value'], instance.inputs['Rotation'])

gn_links.new(instance.outputs['Instances'], set_mat.inputs['Geometry'])

gn_links.new(group_input.outputs['Geometry'], join.inputs['Geometry'])
gn_links.new(set_mat.outputs['Geometry'], join.inputs['Geometry'])
gn_links.new(join.outputs['Geometry'], group_output.inputs['Geometry'])

# Link the shared scattering system modifier to BOTH grass segments exclusively
for target_meadow in [meadow_near, meadow_far]:
    geo_modifier = target_meadow.modifiers.new(name="Grass_Scatter_System", type='NODES')
    geo_modifier.node_group = gn_tree

# Boulders repositioned along the safe visual margins
boulder_configs = [
    (-9.5, 15, 1.0, (4.5, 3.5, 3.8)),
    (-14.0, 28, 2.5, (7.5, 6.0, 7.0)),
    (9.5, 14, 1.0, (4.0, 4.0, 3.5)),
    (14.0, 25, 2.0, (6.0, 5.5, 5.5))
]
boulders = []

for i, (x, y, z, scale_xyz) in enumerate(boulder_configs):
    bpy.ops.mesh.primitive_ico_sphere_add(subdivisions=3, radius=1.0, location=(x, y, z))
    boulder = bpy.context.object
    boulder.name = f"Peripheral_Boulder_{i}"
    boulder.scale = scale_xyz
    boulder.data.materials.append(boulder_mat)
    boulders.append(boulder)

# ==========================================
# 3. ROCK LOOK GENERATOR (MONSTER MATERIAL)
# ==========================================
rock_mat = bpy.data.materials.new(name="Rock_Material")
rock_mat.use_nodes = True
r_nodes = rock_mat.node_tree.nodes
r_links = rock_mat.node_tree.links
r_nodes.clear()

node_r_output = r_nodes.new(type='ShaderNodeOutputMaterial')
node_r_principled = r_nodes.new(type='ShaderNodeBsdfPrincipled')
node_r_principled.inputs['Base Color'].default_value = (0.72, 0.28, 0.06, 1.0)
node_r_principled.inputs['Roughness'].default_value = 0.98

node_r_noise = r_nodes.new(type='ShaderNodeTexNoise')
node_r_noise.inputs['Scale'].default_value = 18.0
node_r_noise.inputs['Detail'].default_value = 8.0

node_r_bump = r_nodes.new(type='ShaderNodeBump')
node_r_bump.inputs['Strength'].default_value = 0.75

r_links.new(node_r_noise.outputs['Fac'], node_r_bump.inputs['Height'])
r_links.new(node_r_bump.outputs['Normal'], node_r_principled.inputs['Normal'])
r_links.new(node_r_principled.outputs['BSDF'], node_r_output.inputs['Surface'])

def apply_rock_material(obj):
    if len(obj.data.materials) == 0:
        obj.data.materials.append(rock_mat)
    else:
        obj.data.materials[0] = rock_mat

monster_parts = []

# ==========================================
# 4. BUILD THE BULKED MONSTER
# ==========================================
root_z_height = 1.9

bpy.ops.object.empty_add(type='PLAIN_AXES', location=(6, 0, root_z_height))
root = bpy.context.object
root.name = "Stick_Root"

torso = bpy.ops.mesh.primitive_cylinder_add(radius=0.32, depth=1.2, location=(0, 0, 0))
torso = bpy.context.object
torso.name = "Torso"
torso.parent = root
torso.location = (0, 0, 0)
bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=3)
bpy.ops.object.mode_set(mode='OBJECT')
apply_rock_material(torso)
monster_parts.append(torso)

bpy.ops.mesh.primitive_uv_sphere_add(radius=0.34, location=(0, 0, 0.85))
head = bpy.context.object
head.name = "Head"
head.parent = torso
apply_rock_material(head)
monster_parts.append(head)

bpy.ops.mesh.primitive_cylinder_add(radius=0.18, depth=0.9, location=(0, 0, 0.5))
shoulders = bpy.context.object
shoulders.name = "Shoulders"
shoulders.parent = torso
shoulders.rotation_euler = (0, math.radians(90), 0)
bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=3)
bpy.ops.object.mode_set(mode='OBJECT')
apply_rock_material(shoulders)
monster_parts.append(shoulders)

bpy.ops.mesh.primitive_cylinder_add(radius=0.15, depth=0.6, location=(0, 0, -0.6))
hips = bpy.context.object
hips.name = "Hips"
hips.parent = torso
hips.rotation_euler = (0, math.radians(90), 0)
bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=3)
bpy.ops.object.mode_set(mode='OBJECT')
apply_rock_material(hips)
monster_parts.append(hips)

def create_limb(name, parent, local_position, length=1.0, radius=0.04):
    bpy.ops.mesh.primitive_cylinder_add(radius=radius, depth=length, location=(0, 0, 0))
    limb = bpy.context.object
    limb.name = name
    bpy.ops.object.mode_set(mode='EDIT')
    bpy.ops.mesh.select_all(action='SELECT')
    bpy.ops.mesh.subdivide(number_cuts=3)
    bpy.ops.transform.translate(value=(0, 0, -length / 2))
    bpy.ops.object.mode_set(mode='OBJECT')
    limb.parent = parent
    limb.location = local_position
    apply_rock_material(limb)
    monster_parts.append(limb)
    return limb

def attach_hand(name, parent, local_position):
    bpy.ops.mesh.primitive_uv_sphere_add(radius=0.16, location=(0, 0, 0))
    hand = bpy.context.object
    hand.name = name
    hand.parent = parent
    hand.location = local_position
    apply_rock_material(hand)
    monster_parts.append(hand)
    return hand

def attach_foot(name, parent, local_position):
    bpy.ops.mesh.primitive_cube_add(size=1, location=(0, 0, 0))
    foot = bpy.context.object
    foot.name = name
    bpy.ops.object.mode_set(mode='EDIT')
    bpy.ops.mesh.select_all(action='SELECT')
    bpy.ops.mesh.subdivide(number_cuts=3)
    bpy.ops.transform.resize(value=(0.34, 0.44, 0.16))
    bpy.ops.transform.translate(value=(0, 0.16, -0.02))
    bpy.ops.object.mode_set(mode='OBJECT')
    foot.parent = parent
    foot.location = local_position
    apply_rock_material(foot)
    monster_parts.append(foot)
    return foot

upper_arm_l = create_limb("Upper_Arm_L", torso, (-0.45, 0, 0.5), length=0.45, radius=0.14)
forearm_l   = create_limb("Forearm_L", upper_arm_l, (0, 0, -0.45), length=0.45, radius=0.12)
hand_l      = attach_hand("Hand_L", forearm_l, (0, 0, -0.45))

upper_arm_r = create_limb("Upper_Arm_R", torso, (0.45, 0, 0.5), length=0.45, radius=0.14)
forearm_r   = create_limb("Forearm_R", upper_arm_r, (0, 0, -0.45), length=0.45, radius=0.12)
hand_r      = attach_hand("Hand_R", forearm_r, (0, 0, -0.45))

leg_segment_length = 0.65
thigh_l = create_limb("Thigh_L", torso, (-0.3, 0, -0.6), length=leg_segment_length, radius=0.16)
shin_l  = create_limb("Shin_L", thigh_l, (0, 0, -leg_segment_length), length=leg_segment_length, radius=0.13)
foot_l  = attach_foot("Foot_L", shin_l, (0, 0, -leg_segment_length))

thigh_r = create_limb("Thigh_R", torso, (0.3, 0, -0.6), length=leg_segment_length, radius=0.16)
shin_r  = create_limb("Shin_R", thigh_r, (0, 0, -leg_segment_length), length=leg_segment_length, radius=0.13)
foot_r  = attach_foot("Foot_R", shin_r, (0, 0, -leg_segment_length))

# ==========================================
# 5. PROP SETUP: SCROLL PIECES
# ==========================================
bpy.ops.mesh.primitive_cylinder_add(radius=0.025, depth=0.45)
scroll_roll_r = bpy.context.object
scroll_roll_r.name = "Scroll_Roll_R"

bpy.ops.mesh.primitive_cylinder_add(radius=0.025, depth=0.45)
scroll_roll_l = bpy.context.object
scroll_roll_l.name = "Scroll_Roll_L"

bpy.ops.mesh.primitive_plane_add(size=1)
scroll_paper = bpy.context.object
scroll_paper.name = "Scroll_Paper"

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.transform.rotate(value=math.radians(90), orient_axis='X')
bpy.ops.transform.resize(value=(1.0, 1.0, 0.36))
bpy.ops.object.mode_set(mode='OBJECT')

mat = bpy.data.materials.new(name="Scroll_Texture_Material")
mat.use_nodes = True
nodes = mat.node_tree.nodes
links = mat.node_tree.links
nodes.clear()

node_tex_coord = nodes.new(type='ShaderNodeTexCoord')
node_mapping   = nodes.new(type='ShaderNodeMapping')
node_texture   = nodes.new(type='ShaderNodeTexImage')
node_principled= nodes.new(type='ShaderNodeBsdfPrincipled')
node_output    = nodes.new(type='ShaderNodeOutputMaterial')

node_mapping.vector_type = 'POINT'
node_texture.interpolation = 'Smart'

image_path = "/Users/dobbinp1/Downloads/IMGP6239_psedit.jpg"
if os.path.exists(image_path):
    try:
        img = bpy.data.images.load(image_path)
        node_texture.image = img
    except Exception as e:
        print(f"Image load error: {e}")
else:
    img = bpy.data.images.new("Fallback_Grid", width=1024, height=1024)
    img.generated_type = 'GRID'
    node_texture.image = img

links.new(node_tex_coord.outputs['UV'], node_mapping.inputs['Vector'])
links.new(node_mapping.outputs['Vector'], node_texture.inputs['Vector'])
links.new(node_texture.outputs['Color'], node_principled.inputs['Base Color'])
links.new(node_principled.outputs['BSDF'], node_output.inputs['Surface'])

scroll_paper.data.materials.append(mat)

# ==========================================
# 6. KINEMATIC TARGET CALCULATIONS
# ==========================================
target_alpha = math.radians(30)
target_gamma = math.radians(30)
eff_length = 0.45 * math.cos(target_alpha) + 0.45 * math.cos(target_alpha + target_gamma)
target_beta_r = math.asin(0.45 / eff_length)
target_beta_l = -target_beta_r

# ==========================================
# 7. ROOT PATH KEYFRAMES
# ==========================================
scene.frame_set(1)
root.location = (6, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(90))
root.keyframe_insert(data_path="location")
root.keyframe_insert(data_path="rotation_euler")

scene.frame_set(65)
root.location = (0, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(90))
root.keyframe_insert(data_path="location")
root.keyframe_insert(data_path="rotation_euler")

scene.frame_set(75)
root.location = (0, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(180))
root.keyframe_insert(data_path="location")
root.keyframe_insert(data_path="rotation_euler")

for f in [92, 105, 125, 145, 160]:
    scene.frame_set(f)
    root.location = (0, 0, root_z_height)
    root.rotation_euler = (0, 0, math.radians(180))
    root.keyframe_insert(data_path="location")
    root.keyframe_insert(data_path="rotation_euler")

# ==========================================
# 8. LIGHTWEIGHT ANIMATION LOOP
# ==========================================
scene.frame_set(125)
bpy.context.view_layer.update()
peak_span = (hand_l.matrix_world.to_translation() - hand_r.matrix_world.to_translation()).length

for frame in range(1, 161):
    scene.frame_set(frame)
    unfurl_factor = 0.0

    if frame <= 65:
        if frame < 55: damping = 1.0
        else: damping = 1.0 - ((frame - 55) / 10.0)

        walk_speed = 0.45
        swing = math.sin(frame * walk_speed) * math.radians(26) * damping
        thigh_cos = math.cos(frame * walk_speed)

        thigh_l.rotation_euler = (swing, 0, 0)
        thigh_r.rotation_euler = (-swing, 0, 0)
        shin_l.rotation_euler = (-max(0.0, thigh_cos) * math.radians(50) * damping, 0, 0)
        shin_r.rotation_euler = (-max(0.0, -thigh_cos) * math.radians(50) * damping, 0, 0)
        foot_l.rotation_euler = (-thigh_l.rotation_euler[0] - shin_l.rotation_euler[0], 0, 0)
        foot_r.rotation_euler = (-thigh_r.rotation_euler[0] - shin_r.rotation_euler[0], 0, 0)

        upper_arm_l.rotation_euler = (-swing, 0, 0)
        forearm_l.rotation_euler = (math.radians(20), 0, 0)

        upper_arm_r.rotation_euler = (math.radians(45), math.radians(-10), 0)
        forearm_r.rotation_euler = (math.radians(60), 0, 0)

    elif frame <= 75:
        t = (frame - 65) / 10.0
        for leg in [thigh_l, thigh_r, shin_l, shin_r, foot_l, foot_r]:
            leg.rotation_euler = (0, 0, 0)

        upper_arm_l.rotation_euler = (math.radians(20) * (1 - t), 0, 0)
        forearm_l.rotation_euler = (math.radians(20) * (1 - t), 0, 0)

        upper_arm_r.rotation_euler = (math.radians(45), math.radians(-10), 0)
        forearm_r.rotation_euler = (math.radians(60), 0, 0)

    elif frame <= 92:
        upper_arm_r.rotation_euler = (math.radians(45), math.radians(-10), 0)
        forearm_r.rotation_euler = (math.radians(60), 0, 0)

        upper_arm_l.rotation_euler = (math.radians(45), math.radians(15), 0)
        beckon_wave = math.sin((frame - 76) * 0.8)
        forearm_l.rotation_euler = (math.radians(60) + beckon_wave * math.radians(25), 0, 0)

    elif frame <= 105:
        t = (frame - 92) / 13.0

        upper_arm_r.rotation_euler = (
            math.radians(45) + t * (target_alpha - math.radians(45)),
            math.radians(-10) + t * (target_beta_r - math.radians(-10)),
            0
        )
        forearm_r.rotation_euler = (math.radians(60) + t * (target_gamma - math.radians(60)), 0, 0)

        start_fore_l = math.radians(60) + math.sin((92 - 76) * 0.8) * math.radians(25)
        upper_arm_l.rotation_euler = (
            math.radians(45) + t * (target_alpha - math.radians(45)),
            math.radians(15) + t * (target_beta_l - math.radians(15)),
            0
        )
        forearm_l.rotation_euler = (start_fore_l + t * (target_gamma - start_fore_l), 0, 0)

    elif frame <= 125:
        t = (frame - 105) / 20.0
        unfurl_factor = t

        curr_beta_r = target_beta_r + t * (math.radians(-15) - target_beta_r)
        curr_beta_l = target_beta_l + t * (math.radians(15) - target_beta_l)

        upper_arm_r.rotation_euler = (target_alpha, curr_beta_r, 0)
        forearm_r.rotation_euler = (target_gamma, 0, 0)

        upper_arm_l.rotation_euler = (target_alpha, curr_beta_l, 0)
        forearm_l.rotation_euler = (target_gamma, 0, 0)

    elif frame <= 145:
        unfurl_factor = 1.0

        upper_arm_r.rotation_euler = (target_alpha, math.radians(-15), 0)
        forearm_r.rotation_euler = (target_gamma, 0, 0)

        upper_arm_l.rotation_euler = (target_alpha, math.radians(15), 0)
        forearm_l.rotation_euler = (target_gamma, 0, 0)

    else:
        t = (frame - 145) / 15.0
        unfurl_factor = 1.0 - t

        curr_beta_r = math.radians(-15) + t * (target_beta_r - math.radians(-15))
        curr_beta_l = math.radians(15) + t * (target_beta_l - math.radians(15))

        upper_arm_r.rotation_euler = (target_alpha, curr_beta_r, 0)
        forearm_r.rotation_euler = (target_gamma, 0, 0)

        upper_arm_l.rotation_euler = (target_alpha, curr_beta_l, 0)
        forearm_l.rotation_euler = (target_gamma, 0, 0)

    for limb in [thigh_l, thigh_r, shin_l, shin_r, foot_l, foot_r, upper_arm_l, forearm_l, upper_arm_r, forearm_r]:
        limb.keyframe_insert(data_path="rotation_euler")

    bpy.context.view_layer.update()

    wpos_r = hand_r.matrix_world.to_translation()
    wpos_l = hand_l.matrix_world.to_translation()

    scroll_roll_r.location = wpos_r
    scroll_roll_l.location = wpos_r + unfurl_factor * (wpos_l - wpos_r)

    scroll_paper.location = (scroll_roll_r.location + scroll_roll_l.location) / 2.0
    span = (scroll_roll_l.location - scroll_roll_r.location).length
    scroll_paper.scale = (max(0.01, span), 1.0, 1.0)

    node_mapping.inputs['Scale'].default_value = (peak_span, 1.0, 1.0)
    node_mapping.inputs['Location'].default_value = (-0.5 * peak_span + 0.5, 0.0, 0.0)

    scroll_roll_r.rotation_euler = (0, 0, 0)
    scroll_roll_l.rotation_euler = (0, 0, 0)
    scroll_paper.rotation_euler = (0, 0, 0)

    for prop in [scroll_roll_r, scroll_roll_l, scroll_paper]:
        prop.keyframe_insert(data_path="location")
        prop.keyframe_insert(data_path="rotation_euler")
    scroll_paper.keyframe_insert(data_path="scale")

    node_mapping.inputs['Scale'].keyframe_insert(data_path="default_value")
    node_mapping.inputs['Location'].keyframe_insert(data_path="default_value")

# ==========================================
# 9. DEFERRED GEOMETRY DECORATION (FAST)
# ==========================================
for part in monster_parts:
    subsurf = part.modifiers.new(name="Rock_Subsurf", type='SUBSURF')
    subsurf.levels = 2
    subsurf.render_levels = 2
    displace = part.modifiers.new(name="Rock_Displace", type='DISPLACE')
    displace.texture = displace_tex
    displace.strength = 0.085

for b in boulders:
    subsurf = b.modifiers.new(name="Boulder_Subsurf", type='SUBSURF')
    subsurf.levels = 1
    subsurf.render_levels = 2
    displace = b.modifiers.new(name="Boulder_Displace", type='DISPLACE')
    displace.texture = boulder_displace_tex
    displace.strength = 0.35
    displace.direction = 'NORMAL'

# ==========================================
# 10. CAMERA MOVEMENT TRACKING
# ==========================================
scene.frame_set(1)
camera.location = (0, -12, 2.5)
camera.keyframe_insert(data_path="location")

scene.frame_set(105)
camera.location = (0, -12, 2.5)
camera.keyframe_insert(data_path="location")

scene.frame_set(125)
camera.location = (0, -5.2, 2.0)
camera.keyframe_insert(data_path="location")

scene.frame_set(1)
print("Physical layout separation complete. Road is perfectly isolated from grass scattering!")
```
