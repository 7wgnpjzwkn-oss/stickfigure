```
import bpy
import math
import mathutils
import os
import random

# ==============================================================================
# 1. ENVIRONMENT SETUP, CLEANUP & CYCLES GPU CONFIG
# ==============================================================================
if bpy.context.object and bpy.context.object.mode != 'OBJECT':
    bpy.ops.object.mode_set(mode='OBJECT')

bpy.ops.object.select_all(action='SELECT')
bpy.ops.object.delete()

scene = bpy.context.scene
scene.frame_start = 1
scene.frame_end = 180
scene.frame_current = 1

# Speed optimization
bpy.context.preferences.edit.use_global_undo = False

# Enable Cycles Add-on & Engine
bpy.ops.preferences.addon_enable(module="cycles")
scene.render.engine = 'CYCLES'
cycles = scene.cycles
cycles.device = 'GPU'

try:
    prefs = bpy.context.preferences.addons['cycles'].preferences
    prefs.get_devices()
    for device_type in ('OPTIX', 'CUDA', 'METAL', 'HIP', 'ONEAPI'):
        try:
            prefs.compute_device_type = device_type
            for dev in prefs.devices:
                dev.use = True
            break
        except Exception:
            pass
except Exception as e:
    print(f"GPU Setup Notice: {e}")

cycles.use_adaptive_sampling = True
cycles.samples = 128
cycles.preview_samples = 32
cycles.use_denoising = True
cycles.denoiser = 'OPENIMAGEDENOISE'


# ==============================================================================
# SHARED HELPER FUNCTIONS
# ==============================================================================
def load_pbr_texture(node_tree, links, vector_output, file_path, is_color=True):
    """Universal texture loader helper to reduce code duplication."""
    node = node_tree.nodes.new(type='ShaderNodeTexImage')
    if os.path.exists(file_path):
        img = bpy.data.images.load(file_path)
        node.image = img
        if not is_color:
            img.colorspace_settings.name = 'Non-Color'
    links.new(vector_output, node.inputs['Vector'])
    return node

def create_or_get_light(name, light_type, location, rotation):
    if name in bpy.data.objects:
        light_obj = bpy.data.objects[name]
    else:
        light_data = bpy.data.lights.new(name=name, type=light_type)
        light_obj = bpy.data.objects.new(name, light_data)
        bpy.context.collection.objects.link(light_obj)

    light_obj.location = location
    light_obj.rotation_euler = rotation
    return light_obj


# ==============================================================================
# 2. CAMERA INITIALIZATION
# ==============================================================================
camera = bpy.data.objects.get("Camera")
if not camera:
    bpy.ops.object.camera_add(location=(0, -12, 2.5), rotation=(math.radians(84), 0, 0))
    camera = bpy.context.object

scene.camera = camera
camera.data.lens = 50
camera.data.dof.use_dof = True
camera.data.dof.aperture_fstop = 2.8

head_obj = bpy.data.objects.get("Scroll_Paper")
camera.data.dof.focus_object = head_obj if head_obj else None
if not head_obj:
    camera.data.dof.focus_distance = 12.0

scene.render.use_motion_blur = True
scene.render.motion_blur_shutter = 0.05


# ==============================================================================
# 3. ENVIRONMENT & SKY LIGHTING (HDRI)
# ==============================================================================
hdri_file_path = "/Volumes/macstorage/Users/dobbinp1/textures/horn-koppe_spring_4k.exr"

world = scene.world or bpy.data.worlds.new("World")
scene.world = world

world.use_nodes = True
wnodes = world.node_tree.nodes
wlinks = world.node_tree.links
wnodes.clear()

node_w_output = wnodes.new(type='ShaderNodeOutputWorld')
node_w_bg = wnodes.new(type='ShaderNodeBackground')
node_w_env = wnodes.new(type='ShaderNodeTexEnvironment')

if os.path.exists(hdri_file_path):
    node_w_env.image = bpy.data.images.load(hdri_file_path)

node_w_coord = wnodes.new(type='ShaderNodeTexCoord')
node_w_mapping = wnodes.new(type='ShaderNodeMapping')

wlinks.new(node_w_coord.outputs['Generated'], node_w_mapping.inputs['Vector'])
wlinks.new(node_w_mapping.outputs['Vector'], node_w_env.inputs['Vector'])
wlinks.new(node_w_env.outputs['Color'], node_w_bg.inputs['Color'])
node_w_bg.inputs['Strength'].default_value = 1.0
wlinks.new(node_w_bg.outputs['Background'], node_w_output.inputs['Surface'])

node_w_mapping.inputs['Rotation'].default_value[2] = math.radians(45.0)

scene.view_settings.view_transform = 'AgX'
scene.view_settings.look = 'AgX - Punchy'
scene.view_settings.exposure = 0


# ==============================================================================
# 4. LIGHT RIG
# ==============================================================================
key_sun = create_or_get_light("Key_Sun", 'SUN', (5, -10, 12), (math.radians(40), math.radians(15), math.radians(-35)))
key_sun.data.color, key_sun.data.energy, key_sun.data.angle = (1.0, 0.88, 0.72), 4.5, math.radians(2.5)

rim_light = create_or_get_light("Rim_Light", 'SUN', (-4, 8, 10), (math.radians(-50), math.radians(-20), math.radians(140)))
rim_light.data.color, rim_light.data.energy, rim_light.data.angle = (0.75, 0.88, 1.0), 6.0, math.radians(1.0)

fill_light = create_or_get_light("Fill_Light", 'SUN', (-8, -5, 6), (math.radians(20), math.radians(-40), math.radians(60)))
fill_light.data.color, fill_light.data.energy, fill_light.data.angle = (0.6, 0.75, 0.95), 1.2, math.radians(10.0)


# ==============================================================================
# 5. PBR GRASS MATERIAL
# ==============================================================================
GRASS_ALBEDO    = "/Volumes/macstorage/Users/dobbinp1/textures/Grass004_4K-JPG/Grass004_4K-JPG_Color.jpg"
GRASS_ROUGHNESS = "/Volumes/macstorage/Users/dobbinp1/textures/Grass004_4K-JPG/Grass004_4K-JPG_Roughness.jpg"
GRASS_NORMAL    = "/Volumes/macstorage/Users/dobbinp1/textures/Grass004_4K-JPG/Grass004_4K-JPG_NormalDX.jpg"
GRASS_DISP      = "/Volumes/macstorage/Users/dobbinp1/textures/Grass004_4K-JPG/Grass004_4K-JPG_Displacement.jpg"

grass_mat = bpy.data.materials.new(name="Grass_Terrain_PBR")
grass_mat.use_nodes = True
grass_mat.displacement_method = 'DISPLACEMENT'

gnodes, glinks = grass_mat.node_tree.nodes, grass_mat.node_tree.links
gnodes.clear()

node_output_g = gnodes.new(type='ShaderNodeOutputMaterial')
node_principled_g = gnodes.new(type='ShaderNodeBsdfPrincipled')
glinks.new(node_principled_g.outputs['BSDF'], node_output_g.inputs['Surface'])

node_coord_g = gnodes.new(type='ShaderNodeTexCoord')
uv_noise_g = gnodes.new(type='ShaderNodeTexNoise')
uv_noise_g.inputs['Scale'].default_value, uv_noise_g.inputs['Detail'].default_value = 1.0, 2.0

uv_distort_mix_g = gnodes.new(type='ShaderNodeMix')
uv_distort_mix_g.data_type = 'RGBA'
uv_distort_mix_g.inputs['Factor'].default_value = 0.06

glinks.new(node_coord_g.outputs['UV'], uv_distort_mix_g.inputs['A'])
glinks.new(uv_noise_g.outputs['Color'], uv_distort_mix_g.inputs['B'])

node_mapping_g = gnodes.new(type='ShaderNodeMapping')
node_mapping_g.inputs['Scale'].default_value = (8.0, 8.0, 1.0)
glinks.new(uv_distort_mix_g.outputs['Result'], node_mapping_g.inputs['Vector'])

map_out_g = node_mapping_g.outputs['Vector']
grass_albedo = load_pbr_texture(grass_mat.node_tree, glinks, map_out_g, GRASS_ALBEDO, True)
grass_rough  = load_pbr_texture(grass_mat.node_tree, glinks, map_out_g, GRASS_ROUGHNESS, False)
grass_normal = load_pbr_texture(grass_mat.node_tree, glinks, map_out_g, GRASS_NORMAL, False)
grass_disp   = load_pbr_texture(grass_mat.node_tree, glinks, map_out_g, GRASS_DISP, False)

macro_noise_g = gnodes.new(type='ShaderNodeTexNoise')
macro_noise_g.inputs['Scale'].default_value, macro_noise_g.inputs['Detail'].default_value = 0.8, 3.0

mix_patch_color = gnodes.new(type='ShaderNodeMix')
mix_patch_color.data_type = 'RGBA'
mix_patch_color.inputs['Factor'].default_value = 0.25
mix_patch_color.inputs['B'].default_value = (0.35, 0.28, 0.15, 1.0)

glinks.new(macro_noise_g.outputs['Fac'], mix_patch_color.inputs['Factor'])
glinks.new(grass_albedo.outputs['Color'], mix_patch_color.inputs['A'])
glinks.new(mix_patch_color.outputs['Result'], node_principled_g.inputs['Base Color'])

mix_roughness_g = gnodes.new(type='ShaderNodeMix')
mix_roughness_g.data_type = 'FLOAT'
mix_roughness_g.inputs['B'].default_value = 0.90

glinks.new(macro_noise_g.outputs['Fac'], mix_roughness_g.inputs['Factor'])
glinks.new(grass_rough.outputs['Color'], mix_roughness_g.inputs['A'])
glinks.new(mix_roughness_g.outputs['Result'], node_principled_g.inputs['Roughness'])

normal_map_node_g = gnodes.new(type='ShaderNodeNormalMap')
normal_map_node_g.inputs['Strength'].default_value = 1.6
glinks.new(grass_normal.outputs['Color'], normal_map_node_g.inputs['Color'])
glinks.new(normal_map_node_g.outputs['Normal'], node_principled_g.inputs['Normal'])

node_disp_g = gnodes.new(type='ShaderNodeDisplacement')
node_disp_g.inputs['Scale'].default_value, node_disp_g.inputs['Midlevel'].default_value = 0.05, 0.5
glinks.new(grass_disp.outputs['Color'], node_disp_g.inputs['Height'])
glinks.new(node_disp_g.outputs['Displacement'], node_output_g.inputs['Displacement'])


# ==============================================================================
# 6. GRAVEL ROAD MESH & MATERIAL
# ==============================================================================
if "Gravel_Road_PBR_Mesh" in bpy.data.objects:
    bpy.data.objects.remove(bpy.data.objects["Gravel_Road_PBR_Mesh"], do_unlink=True)

bpy.ops.mesh.primitive_plane_add(size=1.0, location=(0, 0, 0.02))
road_obj = bpy.context.active_object
road_obj.name = "Gravel_Road_PBR_Mesh"
road_obj.scale = (1000.0, 4.0, 1.0)
bpy.ops.object.transform_apply(scale=True)

bpy.ops.object.mode_set(mode='EDIT')
bpy.ops.mesh.select_all(action='SELECT')
bpy.ops.mesh.subdivide(number_cuts=64)
bpy.ops.uv.unwrap(method='ANGLE_BASED', margin=0.0)
bpy.ops.object.mode_set(mode='OBJECT')

subsurf_mod = road_obj.modifiers.new(name="PBR_Subsurf", type='SUBSURF')
subsurf_mod.subdivision_type = 'CATMULL_CLARK'
subsurf_mod.use_adaptive_subdivision = True
subsurf_mod.adaptive_pixel_size = 1.0

GRAVEL_ALBEDO    = "/Volumes/macstorage/Users/dobbinp1/textures/gravelly_sand_4k.blend/textures/gravelly_sand_diff_4k.jpg"
GRAVEL_ROUGHNESS = "/Volumes/macstorage/Users/dobbinp1/textures/gravelly_sand_4k.blend/textures/gravelly_sand_rough_4k.exr"
GRAVEL_NORMAL    = "/Volumes/macstorage/Users/dobbinp1/textures/gravelly_sand_4k.blend/textures/gravelly_sand_nor_gl_4k.exr"
GRAVEL_DISP      = "/Volumes/macstorage/Users/dobbinp1/textures/gravelly_sand_4k.blend/textures/gravelly_sand_disp_4k.png"

road_mat = bpy.data.materials.new(name="Gravel_Road_PBR_Displaced")
road_mat.use_nodes = True
road_mat.displacement_method = 'DISPLACEMENT'

rnodes, rlinks = road_mat.node_tree.nodes, road_mat.node_tree.links
rnodes.clear()

output_node_r = rnodes.new(type='ShaderNodeOutputMaterial')
principled_r  = rnodes.new(type='ShaderNodeBsdfPrincipled')
rlinks.new(principled_r.outputs['BSDF'], output_node_r.inputs['Surface'])

tex_coord_r = rnodes.new(type='ShaderNodeTexCoord')
uv_noise_r = rnodes.new(type='ShaderNodeTexNoise')
uv_noise_r.inputs['Scale'].default_value, uv_noise_r.inputs['Detail'].default_value = 2.0, 2.0

uv_distort_mix_r = rnodes.new(type='ShaderNodeMix')
uv_distort_mix_r.data_type = 'RGBA'
uv_distort_mix_r.inputs['Factor'].default_value = 0.04

rlinks.new(tex_coord_r.outputs['UV'], uv_distort_mix_r.inputs['A'])
rlinks.new(uv_noise_r.outputs['Color'], uv_distort_mix_r.inputs['B'])

mapping_micro_r = rnodes.new(type='ShaderNodeMapping')
mapping_micro_r.inputs['Scale'].default_value = (50.0, 1.0, 1.0)
rlinks.new(uv_distort_mix_r.outputs['Result'], mapping_micro_r.inputs['Vector'])

map_out_r = mapping_micro_r.outputs['Vector']
road_albedo = load_pbr_texture(road_mat.node_tree, rlinks, map_out_r, GRAVEL_ALBEDO, True)
road_rough  = load_pbr_texture(road_mat.node_tree, rlinks, map_out_r, GRAVEL_ROUGHNESS, False)
road_normal = load_pbr_texture(road_mat.node_tree, rlinks, map_out_r, GRAVEL_NORMAL, False)
road_disp   = load_pbr_texture(road_mat.node_tree, rlinks, map_out_r, GRAVEL_DISP, False)

macro_noise_r = rnodes.new(type='ShaderNodeTexNoise')
macro_noise_r.inputs['Scale'].default_value, macro_noise_r.inputs['Detail'].default_value = 1.8, 4.0

mix_dirt_color = rnodes.new(type='ShaderNodeMix')
mix_dirt_color.data_type = 'RGBA'
mix_dirt_color.inputs['Factor'].default_value = 0.30
mix_dirt_color.inputs['B'].default_value = (0.22, 0.16, 0.11, 1.0)

rlinks.new(macro_noise_r.outputs['Fac'], mix_dirt_color.inputs['Factor'])
rlinks.new(road_albedo.outputs['Color'], mix_dirt_color.inputs['A'])
rlinks.new(mix_dirt_color.outputs['Result'], principled_r.inputs['Base Color'])

mix_roughness_r = rnodes.new(type='ShaderNodeMix')
mix_roughness_r.data_type = 'FLOAT'
mix_roughness_r.inputs['B'].default_value = 0.92

rlinks.new(macro_noise_r.outputs['Fac'], mix_roughness_r.inputs['Factor'])
rlinks.new(road_rough.outputs['Color'], mix_roughness_r.inputs['A'])
rlinks.new(mix_roughness_r.outputs['Result'], principled_r.inputs['Roughness'])

normal_map_node_r = rnodes.new(type='ShaderNodeNormalMap')
normal_map_node_r.inputs['Strength'].default_value = 2.0
rlinks.new(road_normal.outputs['Color'], normal_map_node_r.inputs['Color'])
rlinks.new(normal_map_node_r.outputs['Normal'], principled_r.inputs['Normal'])

node_disp_r = rnodes.new(type='ShaderNodeDisplacement')
node_disp_r.inputs['Scale'].default_value, node_disp_r.inputs['Midlevel'].default_value = 0.08, 0.5
rlinks.new(road_disp.outputs['Color'], node_disp_r.inputs['Height'])
rlinks.new(node_disp_r.outputs['Displacement'], output_node_r.inputs['Displacement'])

road_obj.data.materials.append(road_mat)


# ==============================================================================
# 7. ENVIRONMENT MESHES & MEADOWS
# ==============================================================================
verts = [(-0.04, 0.0, 0.0), (0.04, 0.0, 0.0), (-0.02, 0.0, 0.25), (0.02, 0.0, 0.25), (0.0, 0.03, 0.5)]
faces = [(0, 1, 3, 2), (2, 3, 4)]

mesh = bpy.data.meshes.new("Grass_Blade_Mesh")
mesh.from_pydata(verts, [], faces)
mesh.update()

blade = bpy.data.objects.new("Grass_Blade_Asset", mesh)
bpy.context.collection.objects.link(blade)
blade.data.materials.append(grass_mat)
blade.hide_render = blade.hide_viewport = True

bpy.ops.mesh.primitive_plane_add(size=1000, location=(0, 0, -0.5))
ground = bpy.context.object
ground.name = "Grass_Terrain_Backdrop"
ground.data.materials.append(grass_mat)

for name, y_loc in [("Grass_Meadow_Near", -29.0), ("Grass_Meadow_Far", 29.0)]:
    bpy.ops.mesh.primitive_plane_add(size=1, location=(0, y_loc, 0))
    meadow = bpy.context.object
    meadow.name = name
    meadow.scale = (45, 54, 1)
    bpy.ops.object.transform_apply(scale=True)
    meadow.data.materials.append(grass_mat)

    subsurf = meadow.modifiers.get("Adaptive_Subsurf") or meadow.modifiers.new("Adaptive_Subsurf", 'SUBSURF')
    subsurf.subdivision_type = 'SIMPLE'
    subsurf.use_adaptive_subdivision = True
    subsurf.adaptive_pixel_size = 1.0  # Fixed target bug


# ==============================================================================
# 8. BOULDER GENERATION
# ==============================================================================
boulder_mat = bpy.data.materials.get("PBR_Boulder_Material") or bpy.data.materials.new(name="PBR_Boulder_Material")
boulder_mat.use_nodes = True
bnodes, blinks = boulder_mat.node_tree.nodes, boulder_mat.node_tree.links
bnodes.clear()

node_output_b = bnodes.new(type='ShaderNodeOutputMaterial')
node_bsdf_b   = bnodes.new(type='ShaderNodeBsdfPrincipled')
node_coords_b = bnodes.new(type='ShaderNodeTexCoord')
node_mapping_b = bnodes.new(type='ShaderNodeMapping')

blinks.new(node_coords_b.outputs['Object'], node_mapping_b.inputs['Vector'])

node_noise_b = bnodes.new(type='ShaderNodeTexNoise')
node_noise_b.inputs['Scale'].default_value, node_noise_b.inputs['Detail'].default_value, node_noise_b.inputs['Roughness'].default_value = 8.0, 15.0, 0.7
blinks.new(node_mapping_b.outputs['Vector'], node_noise_b.inputs['Vector'])

node_ramp_color = bnodes.new(type='ShaderNodeValToRGB')
node_ramp_color.color_ramp.elements[0].position, node_ramp_color.color_ramp.elements[0].color = 0.2, (0.05, 0.05, 0.05, 1.0)
node_ramp_color.color_ramp.elements[1].position, node_ramp_color.color_ramp.elements[1].color = 0.8, (0.25, 0.23, 0.20, 1.0)

blinks.new(node_noise_b.outputs['Factor'], node_ramp_color.inputs['Fac'])
blinks.new(node_ramp_color.outputs['Color'], node_bsdf_b.inputs['Base Color'])
node_bsdf_b.inputs['Roughness'].default_value = 0.85

node_bump_b = bnodes.new(type='ShaderNodeBump')
node_bump_b.inputs['Strength'].default_value, node_bump_b.inputs['Distance'].default_value = 0.4, 0.1

blinks.new(node_noise_b.outputs['Factor'], node_bump_b.inputs['Height'])
blinks.new(node_bump_b.outputs['Normal'], node_bsdf_b.inputs['Normal'])
blinks.new(node_bsdf_b.outputs['BSDF'], node_output_b.inputs['Surface'])

for obj in list(bpy.data.objects):
    if obj.name.startswith("Peripheral_Boulder_"):
        bpy.data.objects.remove(obj, do_unlink=True)

boulder_configs = [
    (-9.5, 15, 1.0, (4.5, 3.5, 3.8)),
    (-14.0, 28, 2.5, (7.5, 6.0, 7.0)),
    (9.5, 14, 1.0, (4.0, 4.0, 3.5)),
    (14.0, 25, 2.0, (6.0, 5.5, 5.5))
]

boulders = []
for i, (x, y, z, scale_xyz) in enumerate(boulder_configs):
    bpy.ops.mesh.primitive_ico_sphere_add(subdivisions=4, radius=1.0, location=(x, y, z))
    boulder = bpy.context.active_object
    boulder.name = f"Peripheral_Boulder_{i}"
    boulder.scale = scale_xyz
    boulder.rotation_euler = (random.uniform(0, math.pi), random.uniform(0, math.pi), random.uniform(0, math.pi))
    bpy.ops.object.transform_apply(rotation=True, scale=True)

    disp_mod = boulder.modifiers.new(name="RockShape_Displace", type='DISPLACE')
    tex = bpy.data.textures.get("Boulder_Noise_Texture") or bpy.data.textures.new("Boulder_Noise_Texture", type='VORONOI')
    tex.noise_scale = 1.2
    disp_mod.texture = tex
    disp_mod.strength = 0.2

    bpy.ops.object.shade_smooth()
    boulder.data.materials.append(boulder_mat)
    boulders.append(boulder)


# ==============================================================================
# 9. MONSTER BASE MATERIAL (BOX MAPPED)
# ==============================================================================
ROCK_ALBEDO    = "/Volumes/macstorage/Users/dobbinp1/textures/dry_river_pebbles_diff_4k.jpg"
ROCK_ROUGHNESS = "/Volumes/macstorage/Users/dobbinp1/textures/dry_river_pebbles_rough_4k.exr"
ROCK_NORMAL    = "/Volumes/macstorage/Users/dobbinp1/textures/dry_river_pebbles_nor_gl_4k.exr"

rock_mat = bpy.data.materials.new(name="Rock_Material_PBR")
rock_mat.use_nodes = True
mr_nodes, mr_links = rock_mat.node_tree.nodes, rock_mat.node_tree.links
mr_nodes.clear()

node_output_mr = mr_nodes.new(type='ShaderNodeOutputMaterial')
node_principled_mr = mr_nodes.new(type='ShaderNodeBsdfPrincipled')
mr_links.new(node_principled_mr.outputs['BSDF'], node_output_mr.inputs['Surface'])

node_coord_mr = mr_nodes.new(type='ShaderNodeTexCoord')
node_mapping_mr = mr_nodes.new(type='ShaderNodeMapping')
node_mapping_mr.inputs['Scale'].default_value = (0.5, 0.5, 0.5)
mr_links.new(node_coord_mr.outputs['Object'], node_mapping_mr.inputs['Vector'])

def add_box_texture_node(file_path, is_color_data=True):
    tex_node = mr_nodes.new(type='ShaderNodeTexImage')
    tex_node.projection, tex_node.projection_blend = 'BOX', 0.2
    if os.path.exists(file_path):
        img = bpy.data.images.load(file_path)
        tex_node.image = img
        if not is_color_data:
            img.colorspace_settings.name = 'Non-Color'
    mr_links.new(node_mapping_mr.outputs['Vector'], tex_node.inputs['Vector'])
    return tex_node

tex_albedo_mr = add_box_texture_node(ROCK_ALBEDO, True)
mr_links.new(tex_albedo_mr.outputs['Color'], node_principled_mr.inputs['Base Color'])

tex_roughness_mr = add_box_texture_node(ROCK_ROUGHNESS, False)
mr_links.new(tex_roughness_mr.outputs['Color'], node_principled_mr.inputs['Roughness'])

tex_normal_mr = add_box_texture_node(ROCK_NORMAL, False)
node_normal_map_mr = mr_nodes.new(type='ShaderNodeNormalMap')
node_normal_map_mr.inputs['Strength'].default_value = 1.5
mr_links.new(tex_normal_mr.outputs['Color'], node_normal_map_mr.inputs['Color'])
mr_links.new(node_normal_map_mr.outputs['Normal'], node_principled_mr.inputs['Normal'])

def apply_rock_material(obj):
    if not obj.data.materials:
        obj.data.materials.append(rock_mat)
    else:
        obj.data.materials[0] = rock_mat


# ==============================================================================
# 10. HUMANOID RIG GEOMETRY
# ==============================================================================
monster_parts = []
root_z_height = 2.12

bpy.ops.object.empty_add(type='PLAIN_AXES', location=(6, 0, root_z_height))
root = bpy.context.object
root.name = "Stick_Root"

bpy.ops.mesh.primitive_ico_sphere_add(subdivisions=3, radius=1.0)
torso = bpy.context.object
torso.name, torso.parent, torso.scale = "Torso", root, (0.42, 0.32, 0.58)
bpy.ops.object.transform_apply(scale=True)
apply_rock_material(torso)
monster_parts.append(torso)

bpy.ops.mesh.primitive_uv_sphere_add(radius=0.30, location=(0, 0, 0.88))
head = bpy.context.object
head.name, head.parent, head.scale = "Head", torso, (1.0, 1.1, 1.15)
apply_rock_material(head)
monster_parts.append(head)

bpy.ops.mesh.primitive_ico_sphere_add(subdivisions=3, radius=1.0)
shoulders = bpy.context.object
shoulders.name, shoulders.parent, shoulders.location, shoulders.scale = "Shoulders", torso, (0, 0, 0.48), (0.58, 0.22, 0.18)
bpy.ops.object.transform_apply(scale=True)
apply_rock_material(shoulders)
monster_parts.append(shoulders)

bpy.ops.mesh.primitive_ico_sphere_add(subdivisions=3, radius=1.0)
hips = bpy.context.object
hips.name, hips.parent, hips.location, hips.scale = "Hips", torso, (0, 0, -0.52), (0.44, 0.28, 0.24)
bpy.ops.object.transform_apply(scale=True)
apply_rock_material(hips)
monster_parts.append(hips)

def create_humanoid_limb(name, parent, local_position, length=1.0, start_radius=0.15):
    bpy.ops.mesh.primitive_ico_sphere_add(subdivisions=3, radius=1.0)
    limb = bpy.context.object
    limb.name = name
    limb.scale = (start_radius, start_radius * 0.9, length * 0.5)
    bpy.ops.object.transform_apply(scale=True)

    bpy.ops.object.mode_set(mode='EDIT')
    bpy.ops.mesh.select_all(action='SELECT')
    bpy.ops.transform.translate(value=(0, 0, -length / 2))
    bpy.ops.object.mode_set(mode='OBJECT')

    limb.parent = parent
    limb.location = local_position
    apply_rock_material(limb)
    monster_parts.append(limb)
    return limb

def attach_hand(name, parent, local_position):
    bpy.ops.mesh.primitive_ico_sphere_add(subdivisions=2, radius=0.13)
    hand = bpy.context.object
    hand.name, hand.parent, hand.location, hand.scale = name, parent, local_position, (0.8, 1.2, 0.6)
    apply_rock_material(hand)
    monster_parts.append(hand)
    return hand

def attach_foot(name, parent, local_position):
    bpy.ops.mesh.primitive_cube_add(size=1)
    foot = bpy.context.object
    foot.name, foot.parent = name, parent
    bpy.ops.object.mode_set(mode='EDIT')
    bpy.ops.mesh.select_all(action='SELECT')
    bpy.ops.transform.resize(value=(0.28, 0.45, 0.18))
    bpy.ops.transform.translate(value=(0, 0.18, -0.09))
    bpy.ops.object.mode_set(mode='OBJECT')
    foot.location = local_position
    apply_rock_material(foot)
    monster_parts.append(foot)
    return foot

upper_arm_l = create_humanoid_limb("Upper_Arm_L", torso, (-0.48, 0, 0.45), length=0.45, start_radius=0.13)
forearm_l   = create_humanoid_limb("Forearm_L", upper_arm_l, (0, 0, -0.45), length=0.45, start_radius=0.10)
hand_l      = attach_hand("Hand_L", forearm_l, (0, 0, -0.45))

upper_arm_r = create_humanoid_limb("Upper_Arm_R", torso, (0.48, 0, 0.45), length=0.45, start_radius=0.13)
forearm_r   = create_humanoid_limb("Forearm_R", upper_arm_r, (0, 0, -0.45), length=0.45, start_radius=0.10)
hand_r      = attach_hand("Hand_R", forearm_r, (0, 0, -0.45))

leg_segment_length = 0.65
thigh_l = create_humanoid_limb("Thigh_L", torso, (-0.26, 0, -0.52), length=leg_segment_length, start_radius=0.15)
shin_l  = create_humanoid_limb("Shin_L", thigh_l, (0, 0, -leg_segment_length), length=leg_segment_length, start_radius=0.12)
foot_l  = attach_foot("Foot_L", shin_l, (0, 0, -leg_segment_length))

thigh_r = create_humanoid_limb("Thigh_R", torso, (0.26, 0, -0.52), length=leg_segment_length, start_radius=0.15)
shin_r  = create_humanoid_limb("Shin_R", thigh_r, (0, 0, -leg_segment_length), length=leg_segment_length, start_radius=0.12)
foot_r  = attach_foot("Foot_R", shin_r, (0, 0, -leg_segment_length))


# ==============================================================================
# 11. SCROLL PROP setup
# ==============================================================================
image_path = "/Users/dobbinp1/Downloads/IMGP6239_psedit.jpg"
img_width, img_height = 1024, 1024

scroll_mat = bpy.data.materials.new(name="Scroll_Texture_Material")
scroll_mat.use_nodes = True
snodes, slinks = scroll_mat.node_tree.nodes, scroll_mat.node_tree.links
snodes.clear()

node_tex_coord_s = snodes.new(type='ShaderNodeTexCoord')
node_mapping_s   = snodes.new(type='ShaderNodeMapping')
node_texture_s   = snodes.new(type='ShaderNodeTexImage')
node_principled_s = snodes.new(type='ShaderNodeBsdfPrincipled')
node_output_s    = snodes.new(type='ShaderNodeOutputMaterial')

node_mapping_s.vector_type = 'POINT'
node_texture_s.interpolation = 'Smart'
node_texture_s.extension = 'CLIP'

if os.path.exists(image_path):
    try:
        img = bpy.data.images.load(image_path)
        node_texture_s.image = img
        img_width, img_height = img.size[0], img.size[1]
    except Exception as e:
        print(f"Image load error: {e}")
else:
    img = bpy.data.images.new("Fallback_Grid", width=img_width, height=img_height)
    img.generated_type = 'GRID'
    node_texture_s.image = img

aspect_ratio = img_height / img_width

bpy.ops.mesh.primitive_cylinder_add(radius=0.025, depth=1.0)
scroll_roll_r = bpy.context.object
scroll_roll_r.name = "Scroll_Roll_R"

bpy.ops.mesh.primitive_cylinder_add(radius=0.025, depth=1.0)
scroll_roll_l = bpy.context.object
scroll_roll_l.name = "Scroll_Roll_L"

bpy.ops.mesh.primitive_plane_add(size=1)
scroll_paper = bpy.context.object
scroll_paper.name = "Scroll_Paper"

slinks.new(node_tex_coord_s.outputs['UV'], node_mapping_s.inputs['Vector'])
slinks.new(node_mapping_s.outputs['Vector'], node_texture_s.inputs['Vector'])
slinks.new(node_texture_s.outputs['Color'], node_principled_s.inputs['Base Color'])
slinks.new(node_principled_s.outputs['BSDF'], node_output_s.inputs['Surface'])
scroll_paper.data.materials.append(scroll_mat)


# ==============================================================================
# 12. ANIMATION SEQUENCING & INVERSE KINEMATICS
# ==============================================================================
target_alpha = math.radians(30)
target_gamma = math.radians(30)
eff_length = 0.45 * math.cos(target_alpha) + 0.45 * math.cos(target_alpha + target_gamma)
target_beta_r = math.asin(0.45 / eff_length)
target_beta_l = -target_beta_r

scene.frame_set(1)
root.location = (6, 0, root_z_height)
root.rotation_euler = (0, 0, math.radians(90))

# PASS 1: Calculate max physical hand span
absolute_max_span = 0.01
for frame in range(1, 180):
    scene.frame_set(frame)
    if frame < 92:
        alpha, gamma = 0, 0
    elif 92 <= frame < 105:
        t = (frame - 92) / (105 - 92)
        alpha, gamma = target_alpha * t, target_gamma * t
    elif 105 <= frame < 145:
        alpha, gamma = target_alpha, target_gamma
    else:
        t = (frame - 145) / (160 - 145)
        alpha, gamma = target_alpha * (1.0 - t), target_gamma * (1.0 - t)

    upper_arm_r.rotation_euler = (alpha, math.radians(-15), 0)
    forearm_r.rotation_euler   = (gamma, 0, 0)
    upper_arm_l.rotation_euler = (alpha, math.radians(15), 0)
    forearm_l.rotation_euler   = (gamma, 0, 0)

    bpy.context.view_layer.update()
    wpos_r = hand_r.matrix_world.to_translation()
    wpos_l = hand_l.matrix_world.to_translation()

    current_physical_distance = (wpos_l - wpos_r).length
    if current_physical_distance > absolute_max_span:
        absolute_max_span = current_physical_distance

# PASS 2: Apply main locomotion & scroll keyframes
walk_start_x, walk_end_x = 6.0, 0.0
total_walk_frames = 64

for frame in range(1, 181):
    scene.frame_set(frame)

    if frame < 92:
        alpha, gamma, beta_r, beta_l = 0, 0, 0, 0
        unfurl_factor = 0.0
    elif 92 <= frame < 105:
        t = (frame - 92) / (105 - 92)
        alpha, gamma = target_alpha * t, target_gamma * t
        beta_r, beta_l = target_beta_r * t, target_beta_l * t
        unfurl_factor = 0.0
    elif 105 <= frame < 125:
        t = (frame - 105) / (125 - 105)
        alpha, gamma = target_alpha, target_gamma
        beta_r = target_beta_r * (1.0 - t)
        beta_l = target_beta_l * (1.0 - t)
        unfurl_factor = t
    else:
        alpha, gamma, beta_r, beta_l = target_alpha, target_gamma, 0, 0
        unfurl_factor = 1.0

    upper_arm_r.rotation_euler = (alpha, math.radians(-15) + beta_r, 0)
    forearm_r.rotation_euler   = (gamma, 0, 0)
    upper_arm_l.rotation_euler = (alpha, math.radians(15) + beta_l, 0)
    forearm_l.rotation_euler   = (gamma, 0, 0)

    if frame < 65:
        walk_speed = 0.5
        stride_amplitude = math.radians(26)
        phase = frame * walk_speed

        progress = (frame - 1) / (total_walk_frames - 1)
        current_x = walk_start_x - (walk_start_x - walk_end_x) * progress

        thigh_l.rotation_euler = (-math.sin(phase) * stride_amplitude, 0, 0)
        thigh_r.rotation_euler = (math.sin(phase) * stride_amplitude, 0, 0)
        shin_l.rotation_euler  = (-max(0, math.cos(phase)) * math.radians(28), 0, 0)
        shin_r.rotation_euler  = (-max(0, -math.cos(phase)) * math.radians(28), 0, 0)
        foot_l.rotation_euler  = (-math.cos(phase) * math.radians(10), 0, 0)
        foot_r.rotation_euler  = (math.cos(phase) * math.radians(10), 0, 0)

        hips.rotation_euler  = (0, 0, math.sin(phase) * math.radians(6))
        torso.rotation_euler = (0, 0, -math.sin(phase) * math.radians(4))

        z_drop = -abs(math.sin(phase)) * 0.06
        root.location = (current_x, 0, root_z_height + z_drop)
        root.rotation_euler = (0, 0, math.radians(90))
    else:
        for limb in [thigh_l, thigh_r, shin_l, shin_r, foot_l, foot_r, hips, torso]:
            limb.rotation_euler = (0, 0, 0)
        root.location = (walk_end_x, 0, root_z_height)
        root.rotation_euler = (0, 0, math.radians(90 if frame < 75 else 180))

    for limb in [root, torso, hips, thigh_l, thigh_r, shin_l, shin_r, foot_l, foot_r, upper_arm_l, forearm_l, upper_arm_r, forearm_r]:
        limb.keyframe_insert(data_path="location")
        limb.keyframe_insert(data_path="rotation_euler")

    bpy.context.view_layer.update()
    wpos_r = hand_r.matrix_world.to_translation()
    wpos_l = hand_l.matrix_world.to_translation()

    wpos_r.y -= 0.15
    wpos_l.y -= 0.15

    scroll_roll_r.location = wpos_r
    scroll_roll_l.location = wpos_r + unfurl_factor * (wpos_l - wpos_r)
    scroll_paper.location  = (scroll_roll_r.location + scroll_roll_l.location) / 2.0

    target_width  = max(0.01, absolute_max_span * unfurl_factor)
    target_height = absolute_max_span * aspect_ratio

    scroll_paper.scale  = (target_width, target_height, 1.0)
    scroll_roll_r.scale = scroll_roll_l.scale = (1.0, 1.0, target_height)

    scroll_roll_r.rotation_euler = scroll_roll_l.rotation_euler = scroll_paper.rotation_euler = (math.radians(90), 0, 0)

    hide_val = unfurl_factor > 0.0
    scroll_roll_r.hide_viewport = scroll_roll_r.hide_render = hide_val
    scroll_roll_l.hide_viewport = scroll_roll_l.hide_render = hide_val

    u_scale = absolute_max_span / target_width if target_width > 0 else 1.0
    node_mapping_s.inputs['Scale'].default_value = (u_scale, 1.0, 1.0)
    node_mapping_s.inputs['Location'].default_value = (-0.5 * u_scale + 0.5, 0.0, 0.0)

    scroll_paper.keyframe_insert(data_path="scale")
    scroll_roll_r.keyframe_insert(data_path="scale")
    scroll_roll_l.keyframe_insert(data_path="scale")

    scroll_roll_r.keyframe_insert(data_path="hide_viewport")
    scroll_roll_r.keyframe_insert(data_path="hide_render")
    scroll_roll_l.keyframe_insert(data_path="hide_viewport")
    scroll_roll_l.keyframe_insert(data_path="hide_render")

    for prop in [scroll_roll_r, scroll_roll_l, scroll_paper]:
        prop.keyframe_insert(data_path="location")
        prop.keyframe_insert(data_path="rotation_euler")

    node_mapping_s.inputs['Scale'].keyframe_insert(data_path="default_value")
    node_mapping_s.inputs['Location'].keyframe_insert(data_path="default_value")


# ==============================================================================
# 13. CHISELED GEOMETRY STACK
# ==============================================================================
rock_displace_tex = bpy.data.textures.new(name="Rock_Micro_Deform", type='VORONOI')
rock_displace_tex.noise_scale = 0.15
rock_displace_tex.distance_metric = 'DISTANCE'

for part in monster_parts:
    bevel = part.modifiers.new(name="Rock_Facet_Edges", type='BEVEL')
    bevel.width, bevel.segments, bevel.limit_method, bevel.angle_limit = 0.03, 2, 'ANGLE', math.radians(30)

    decimate = part.modifiers.new(name="Rock_Chisel", type='DECIMATE')
    decimate.decimate_type, decimate.ratio = 'COLLAPSE', 0.45

    displace = part.modifiers.new(name="Rock_Displace", type='DISPLACE')
    displace.texture, displace.strength = rock_displace_tex, 0.04

for b in boulders:
    for poly in b.data.polygons:
        poly.use_smooth = True


# ==============================================================================
# 14. CAMERA KEYFRAMES & SHOT ALIGNMENT
# ==============================================================================
camera.animation_data_clear()

cam_keyframes = [
    (1, (0, -12, 2.5), (math.radians(84), 0, 0)),
    (105, (0, -12, 2.5), (math.radians(84), 0, 0)),
    (125, (0, -5.2, 2.0), (math.radians(84), 0, 0)),
    (145, (0, -5.2, 2.0), (math.radians(84), 0, 0))
]

for frame, loc, rot in cam_keyframes:
    scene.frame_set(frame)
    camera.location, camera.rotation_euler = loc, rot
    camera.keyframe_insert(data_path="location")
    camera.keyframe_insert(data_path="rotation_euler")

scene.frame_set(145)
bpy.context.view_layer.update()
scroll_world_pos = scroll_paper.matrix_world.to_translation()

scene.frame_set(180)
camera.location = (scroll_world_pos.x, scroll_world_pos.y - 2.0, scroll_world_pos.z)
camera.rotation_euler = (math.radians(90), 0, 0)
camera.keyframe_insert(data_path="location")
camera.keyframe_insert(data_path="rotation_euler")

scene.frame_set(1)


# ==============================================================================
# 15. COMPOSITOR SETUP
# ==============================================================================
# 15. COMPOSITOR SETUP (Blender 5.0+ API)
# ==============================================================================
scene.render.use_compositing = True

# 1. Create a generic node group for compositing and assign it to the scene
comp_tree = bpy.data.node_groups.new(name="MainCompositor", type="CompositorNodeTree")
scene.compositing_node_group = comp_tree

# 2. Add the Render Layers node
comp_rl = comp_tree.nodes.new(type='CompositorNodeRLayers')

# 3. Add the Group Output node (Replaces 'CompositorNodeComposite')
comp_output = comp_tree.nodes.new(type='NodeGroupOutput')

# 4. Explicitly create the 'Image' socket on the tree's interface level
comp_tree.interface.new_socket(name="Image", in_out='OUTPUT', socket_type='NodeSocketColor')

# 5. Link Render Layers to the Output
comp_tree.links.new(comp_rl.outputs['Image'], comp_output.inputs['Image'])

for area in bpy.context.window.screen.areas:
    if area.type == 'VIEW_3D':
        for space in area.spaces:
            if space.type == 'VIEW_3D':
                space.shading.type = 'RENDERED'
                space.shading.use_compositor = 'ALWAYS'
                space.region_3d.view_perspective = 'CAMERA'
```

```
import os
import threading
import asyncio
import numpy as np
from scipy.io import wavfile
from scipy.signal import resample
import edge_tts
import soundfile as sf

def fetch_voice_in_thread(text, output_filename):
    """
    Creates an isolated loop on a background thread.
    This bypasses Blender's restrictive application event loop.
    """
    async def amain():
        VOICE = "en-US-ChristopherNeural"
        communicate = edge_tts.Communicate(text, VOICE)
        await communicate.save(output_filename)

    # Spin up an independent loop just for this thread
    new_loop = asyncio.new_event_loop()
    try:
        new_loop.run_until_complete(amain())
    finally:
        new_loop.close()

def generate_rock_monster_speech(text, output_filename="rock_monster.wav"):
    """Applies the heavy, gravelly, rock-monster audio manipulation."""
    temp_file = "temp_voice.mp3"

    print("рџ¤– Fetching deep base voice via background thread...")

    # Run the network-dependent task inside a separate thread
    worker_thread = threading.Thread(
        target=fetch_voice_in_thread,
        args=(text, temp_file)
    )
    worker_thread.start()
    worker_thread.join()  # Wait for the network task to finish before proceeding

    if not os.path.exists(temp_file):
        raise FileNotFoundError("Failed to download the base voice file.")

    print("рџ”Љ Reading downloaded audio file...")
    data, sample_rate = sf.read(temp_file)

    # If stereo, convert to mono for uniform processing
    if len(data.shape) > 1:
        data = data[:, 0]

    print("рџЄЁ Applying rock monster vocal effects...")

    # Effect 1: Pitch down the voice significantly
    pitch_factor = 0.60  # Lower = deeper and heavier
    num_samples = int(len(data) / pitch_factor)
    deep_data = resample(data, num_samples)

    # Effect 2: Apply soft-clipping distortion to add a gravelly, rocky texture
    max_val = np.max(np.abs(deep_data))
    normalized = deep_data / max_val
    distorted = np.tanh(normalized * 3.5) # Raised to 3.5 for a rockier texture

    # Effect 3: Create a metallic, cavernous resonance (subtle short echo)
    delay_samples = int(sample_rate * 0.035) # 35ms cave echo delay
    echo_data = np.zeros(len(distorted) + delay_samples)
    echo_data[:len(distorted)] += distorted
    echo_data[delay_samples:] += distorted * 0.50

    # Re-normalize to prevent clipping and convert to 16-bit integers
    final_audio = echo_data / np.max(np.abs(echo_data))
    final_audio = (final_audio * 32767).astype(np.int16)

    # Save final audio file
    wavfile.write(output_filename, sample_rate, final_audio)

    # Clean up the temporary base file
    if os.path.exists(temp_file):
        os.remove(temp_file)

    print(f"вњ… Success! Rock monster audio saved to: {os.path.abspath(output_filename)}")

if __name__ == "__main__":
    monster_text = "Who dares awaken the ancient stone? Tremble before the mountain!"
    generate_rock_monster_speech(monster_text)
```
